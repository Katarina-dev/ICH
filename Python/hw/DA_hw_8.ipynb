```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('fmri')
df.head()
df_timepoint_5 = df[df['timepoint'] == 5]
sns.barplot(x='timepoint', y='signal', data=df_timepoint_5, hue='event')
plt.title("Average signal for events in timepoint=5")
plt.show()
```

```python
sns.pointplot(data=df, x='timepoint', y='signal', hue='event')
plt.title("Average signal for events in every timepoint")
plt.show()
```
