```python
Напишите программу, которая:
● Запрашивает у пользователя количество точек данных N.
● Генерирует массив из N случайных точек (координаты x и y), используя нормальное распределение.
● Выводит на экран первые 10 сгенерированных точек (если N было больше 10).
● Вычисляет и выводит среднее значение и стандартное отклонение для координат x и y.
```

```python
N = int(input('Count:'))

n_arr = np.random.normal(size = (N, 2))
print(n_arr)

print("-------")

mean_arr = n_arr.mean(axis=0)
print(mean_arr)

avg_x = n_arr.mean(axis=0)[0]
print(avg_x)

avg_y = n_arr.mean(axis=0)[1]
print(avg_y)

std_arr = n_arr.std(axis=0)
print(std_arr)

std_x = std_arr[0]
print(std_x)

std_y = std_arr[1]
print(std_y)
```

```python
N = int(input("количество точек: "))
n_arr = np.random.normal(size=(N, 2))
print(n_arr)
print("-------")
mean_arr = n_arr.mean(axis=0)
av_x = mean_arr[0]
print(av_x)
av_y = mean_arr[1]
print(av_y)
print("-------")
std_arr = n_arr.std(axis=0)
std_x = std_arr[0]
print(std_x)
std_y = std_arr[1]
print(std_y)
```
