## Работа с Duplicate

```python
duplicates_id = deals_df_v2[deals_df_v2.duplicated(subset=['Id'])]
duplicates_id
```

```python
duplicates_cm = deals_df_v2[deals_df_v2.duplicated(subset=['Contact Name'])]
duplicates_cm
```

```python
duplicates_all = deals_df_v2[deals_df_v2.duplicated(keep=False)]
duplicates_all
```

```python
df_grouped = (
    deals_df_v2
    .groupby('Contact Name')
    .agg({
        'Stage': lambda x: ', '.join(sorted(x.dropna().unique())),
        'Lost Reason': lambda x: ', '.join(sorted(x.dropna().unique()))
    })
    .reset_index()
)
```

```python
#1. Отфильтруем только нужные строки:
filtered_df = deals_df_v2[
    (deals_df_v2['Stage'] == 'Lost') &
    (deals_df_v2['Lost Reason'] == 'Duplicate') &
    (deals_df_v2['Contact Name'].notna())
]
filtered_df
```

```python
#2. Посмотрим, какие Contact Name встречаются более одного раза:
duplicate_count = filtered_df['Contact Name'].value_counts()
duplicate_count
```

```python
duplicate_ids_more_one = duplicate_count[duplicate_count > 1].index.tolist()
duplicate_ids_more_one
```

```python
len(duplicate_ids_more_one)
```

```python
duplicates_contact_name = filtered_df[filtered_df['Contact Name'].isin(duplicate_ids_more_one)]
duplicates_contact_name
```

```python
duplicates_contact_name['Product'].notna().sum()
```

```python
duplicates_contact_name[duplicates_contact_name['Product'].notna()]
```

```python
duplicates_contact_name[duplicates_contact_name['Contact Name']=='5805028000056392363']
```

```python
#3. Оставим только одну строку на каждый дублирующийся ID — самую свежую:
latest_duplicates_contact = (
    duplicates_contact_name
    .sort_values('Created Time')
    .groupby('Contact Name', as_index=False)
    .last()
)
```

```python
latest_duplicates_contact[latest_duplicates_contact['Contact Name']=='5805028000056392363']
```

```python
latest_duplicates_contact
```

```python
import numpy as np

aggregated_contacts = (
    duplicates_contact_name
    .sort_values('Created Time')
    .groupby('Contact Name')
    .agg(lambda x: x.dropna().iloc[-1] if x.dropna().size > 0 else np.nan)
    .reset_index()
)
```

```python
aggregated_contacts
```

```python
aggregated_contacts[aggregated_contacts['Contact Name']=='5805028000056392363']
```

```python
# 5. Приводим порядок колонок в aggregated_contacts к deals_df
aggregated_contacts = aggregated_contacts[deals_df_v2.columns]
aggregated_contacts.head()
```

```python
aggregated_contacts[aggregated_contacts['Contact Name']=='5805028000000872003']
```

```python
# 6. Удаляем старые строки с этими Contact Name
mask_to_remove = (
    (deals_df_v2['Stage'] == 'Lost') &
    (deals_df_v2['Lost Reason'] == 'Duplicate') &
    (deals_df_v2['Contact Name'].isin(duplicate_ids_more_one))
)
deals_df_v2 = deals_df_v2[~mask_to_remove].copy()
deals_df_v2.shape[0]
```

```python
deals_df_v2[deals_df_v2['Contact Name']=='5805028000056392363']
```

```python
# # 7. Добавляем очищенные агрегированные строки.  Склеиваем без сброса индексов
deals_df_v2 = pd.concat([deals_df_v2, aggregated_contacts], ignore_index=True)
```

```python
# 1. Список Contact Name, которые мы удалили:
removed_contact_names = deals_df_v2[
    (deals_df_v2['Stage'] == 'Lost') &
    (deals_df_v2['Lost Reason'] == 'Duplicate') &
    (deals_df_v2['Contact Name'].isin(duplicate_ids_more_one))
]['Contact Name'].unique()
```

```python
len(removed_contact_names)
```

```python
#2. Список Contact Name, которые вернулись через aggregated_contacts
aggregated_contact_names = aggregated_contacts['Contact Name'].unique()
len(aggregated_contact_names)
```

```python
# 3. Сравниваем, кого не хватает:
missing_contacts = set(removed_contact_names) - set(aggregated_contact_names)
len(missing_contacts)  # Сколько потеряно?
```
