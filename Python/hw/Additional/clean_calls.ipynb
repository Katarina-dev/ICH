```python
# Получаем моду (самое частое значение)
mode_value = calls_df['Scheduled in CRM'].mode(dropna=True)[0]

# Заполняем пропуски этим значением
calls_df['Scheduled in CRM'] = calls_df['Scheduled in CRM'].fillna(mode_value)
calls_df.isnull().sum()
```

```python
# Проверим строки, где есть пропущенные значения в столбце 'Scheduled in CRM'
missing_values = calls_df[calls_df['Outgoing Call Status'].isna()]

# Проверим, является ли 'Call Type' для этих строк 'Incoming'
incoming_calls_with_missing = missing_values[missing_values['Call Type'].isin(['Inbound', 'Missed'])]

# Выведем количество таких строк
print(f"Количество пропущенных значений в 'Outgoing Call Status' для входящих звонков: {incoming_calls_with_missing.shape[0]}")
```
