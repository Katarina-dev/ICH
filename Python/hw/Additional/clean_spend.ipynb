```python
import pandas as pd

# Загружаем Excel
excel_file_spend = pd.read_excel('Spend (Done).xlsx')

excel_file_spend.to_csv("Spend.csv", index=False, encoding='utf-8-sig')

spend_df = pd.read_csv("Spend.csv")

spend_df
```

```python
# Заменяет ячейки, которые:
# - полностью пустые ('')
# - содержат только пробелы (' ')
# на NaN.
import numpy as np

spend_df = spend_df.replace(r'^\s*$', np.nan, regex=True)
spend_df.isnull().sum()
```

```python
# Найдём дубликаты
duplicates = spend_df[spend_df.duplicated(keep=False)]

# Сохраняем в Excel
duplicates.to_excel("duplicates_spend.xlsx", index=False)
```

```python
spend_df = spend_df.drop_duplicates(keep='first')
```

```python
spend_df
```

```python
# 3. Удаляем строки, где кроме 'Date' и 'Source' всё остальное — пустое
empty = spend_df.drop(columns=['Date', 'Source']).isna().all(axis=1)
only_date_source = spend_df[empty]  # Сохраним на случай, если пригодится
spend_df = spend_df[~empty]         # Удалим из основного датафрейма
spend_df
```

```python
spend_df.isnull().sum()
```

```python
spend_df[spend_df.duplicated()]
```

```python
# Оставляем строки, где хотя бы одна метрика не равна нулю
spend_df = spend_df[
    (spend_df['Spend'] != 0) | 
    (spend_df['Clicks'] != 0) | 
    (spend_df['Impressions'] != 0)
]
spend_df
```

```python
spend_df.isnull().sum()
```

```python
spend_df.to_csv("cleaned_spend.csv", index=False)
```

```python
spend_df.loc[:, 'Source'] = spend_df['Source'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
spend_df.loc[:, 'Campaign'] = spend_df['Campaign'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
spend_df.loc[:, 'AdGroup'] = spend_df['AdGroup'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
spend_df.loc[:, 'Ad'] = spend_df['Ad'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
spend_df
```

```python
spend_df.nunique()
```

```python
spend_df.loc[:, 'Source'] = spend_df['Source'].astype('category')
spend_df.loc[:, 'Campaign'] = spend_df['Campaign'].astype('category')
spend_df.loc[:, 'AdGroup'] = spend_df['AdGroup'].astype('category')
spend_df.dtypes
```

```python
spend_df.info()
```

```python
spend_df.loc[:, ['Source', 'Campaign', 'AdGroup']] = spend_df[['Source', 'Campaign', 'AdGroup']].apply(
    lambda col: col.cat.add_categories('No data') if 'No data' not in col.cat.categories else col).fillna('No data')
spend_df
```

```python
spend_df.dtypes
```

```python
spend_df.loc[:,'Ad'] = spend_df['Ad'].fillna('No data')
spend_df
```
