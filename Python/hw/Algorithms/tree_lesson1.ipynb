```python
def delete_deepest(root, d_node):
    q = []
    q.append(root)
    
    while q:
        temp = q.pop(0)
        
        if temp.left:
            if temp.left is d_node:
                temp.left = None
                return
            else:
                q.append(temp.left)
                
        if temp.right:
            if temp.right is d_node:
                temp.right = None
                return
            else:
                q.append(temp.right)
```

```python
def delete_node(root, val):
    
    if root is None:
        return root
    
    if val < root.data:     # значение которое ищем меньше значени корня
        root.left = delete_node(root.left, val)
        
    elif val > root.data:   # значение которое ищем больше значени корня
        root.right = delete_node(root.right, val)
    
    else:
        
        if root.right is None and root.left is None:  # если правый и левый дочерние узлы - отсутствуют
            return None
        
        if root.left is None:   # если отсутствует только правый дочерний узел
            return root.right
        
        elif root.right is None:  # отсутствует только левый
            return root.left
        
        temp = find_min(root.right)    # в ином случае ищем в правом поддереве (от узла который хотим удалить),
        root.data = temp.data          # узел с самым маленьким значением
        root.right = delete_node(root.right, temp.data) 
        
    return root
    
    
        
        
        
def find_min(node):
    current = node
    while current.left is not None:   # эквивалентно просто  while current.left
        current = current.left
    return current



        
        
    
        
        
    
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None
```

```python
def sum_tree(root):
    if root is None:
        return 0
    else:
        return sum_tree(root.left) + root.data + sum_tree(root.right) # сумма левого поддерева, корня и правого поддерева
    
def is_sum_tree(node):
    if node is None or (node.left is None and node.right is None): # если корень пустой, или у корня нет потомков
        return 1
    
    ls = sum_tree(node.left)   
    rs = sum_tree(node.right)
    
    
    if (node.data == ls + rs) and is_sum_tree(node.left) and is_sum_tree(node.right):  # если значение корня яаляется суммой значений всех потомков левого и правого поддерева и правое и левое поддерево тоже являются таковыми
        return 1
    else:
        return 0
    
```

```python
class TreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def inorder(root):
    if not root:
        return []
    return inorder(root.left) +  [root.data] + inorder(root.right)

def preorder(root):
    if not root:
        return []
    return [root.data] + inorder(root.left) + inorder(root.right)

def postorder(root):
    if not root:
        return []
    return inorder(root.left) + inorder(root.right) + [root.data]

root = TreeNode(10)
root.left = TreeNode(5)
root.right = TreeNode(20)
root.left.left = TreeNode(3)
root.left.right = TreeNode(7)
root.right.right = TreeNode(30)

print(inorder(root))
print(preorder(root))
print(postorder(root))
```
