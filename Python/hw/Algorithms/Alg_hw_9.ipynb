```python
Реализовать методы в MyLinkedList:

pushToTail(int data), 

removeLast(), 

get(int index)
```

```python
class Node:  # Класс, отвечающий за один узел
    def __init__(self, value=None, next=None):
        self.value = value
        self.next = next


class MyLinkedList:  # Класс, реализующий функциональность списка
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def push_to_head(self, data):  # Добавляет элемент в начало списка
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
        if self.size == 0:
            self.tail = new_node
        self.size += 1

    def push_to_tail(self, data):  # Добавляет элемент в конец списка
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.size += 1

    def push_to_index(self, index, data):  # Вставляет элемент по индексу
        if index < 0 or index > self.size:
            raise IndexError('index out of range')

        if index == 0:
            self.push_to_head(data)
            return

        if index == self.size:
            self.push_to_tail(data)
            return

        new_node = Node(data)
        current = self.head
        for _ in range(index - 1):
            current = current.next

        new_node.next = current.next
        current.next = new_node
        self.size += 1

    def remove_first(self):  # Удаляет первый элемент
        if self.head is None:
            raise IndexError('список пуст')
        self.head = self.head.next
        self.size -= 1
        if self.size == 0:
            self.tail = None

    def remove_last(self):  # Удаляет последний элемент
        if self.head is None:
            raise IndexError('список пуст')
        if self.head.next is None:
            self.head = None
            self.tail = None
        else:
            current = self.head
            while current.next.next:
                current = current.next
            current.next = None
            self.tail = current
        self.size -= 1

    def remove_to_index(self, index):  # Удаляет элемент по индексу
        if index < 0 or index >= self.size:
            raise IndexError('index out of range')

        if index == 0:
            self.remove_first()
            return

        current = self.head
        for _ in range(index - 1):
            current = current.next

        current.next = current.next.next
        if index == self.size - 1:
            self.tail = current
        self.size -= 1

    def get_index(self, index):  # Возвращает значение по индексу
        if index < 0 or index >= self.size:
            raise IndexError('index out of range')
        current = self.head
        for _ in range(index):
            current = current.next
        return current.value

    def print_linked_list(self):  # Печатает весь список
        current = self.head
        elements = []
        while current:
            elements.append(current.value)
            current = current.next
        print(" -> ".join(map(str, elements)))
```

```python
mll = MyLinkedList()  
mll.push_to_head(1)
mll.push_to_head(14)
mll.push_to_head(10)
mll.push_to_head(15)
mll.push_to_head(30)
mll.push_to_head(21)
mll.push_to_tail(222)

mll.print_linked_list()  
print("Размер списка:", mll.size)  
print("Элемент по индексу 3:", mll.get_index(3)) 

mll.remove_last()
mll.print_linked_list()
```

```python
class Node:  # Узел с двумя связями: вперёд и назад
    def __init__(self, value=None, next=None, prev=None):
        self.value = value
        self.next = next
        self.prev = prev


class MyDoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
        self.size = 0

    def push_to_head(self, data):
        new_node = Node(data)
        new_node.next = self.head
        if self.head:
            self.head.prev = new_node
        else:
            self.tail = new_node  # если список пустой
        self.head = new_node
        self.size += 1

    def push_to_tail(self, data):
        new_node = Node(data)
        new_node.prev = self.tail
        if self.tail:
            self.tail.next = new_node
        else:
            self.head = new_node  # если список пустой
        self.tail = new_node
        self.size += 1

    def push_to_index(self, index, data):
        if index < 0 or index > self.size:
            raise IndexError('index out of range')

        if index == 0:
            self.push_to_head(data)
            return
        if index == self.size:
            self.push_to_tail(data)
            return

        new_node = Node(data)
        current = self.head
        for _ in range(index):
            current = current.next

        prev_node = current.prev
        new_node.prev = prev_node
        new_node.next = current
        prev_node.next = new_node
        current.prev = new_node
        self.size += 1

    def remove_first(self):
        if self.head is None:
            raise IndexError("список пуст")
        self.head = self.head.next
        if self.head:
            self.head.prev = None
        else:
            self.tail = None
        self.size -= 1

    def remove_last(self):
        if self.tail is None:
            raise IndexError("список пуст")
        self.tail = self.tail.prev
        if self.tail:
            self.tail.next = None
        else:
            self.head = None
        self.size -= 1

    def remove_to_index(self, index):
        if index < 0 or index >= self.size:
            raise IndexError('index out of range')

        if index == 0:
            self.remove_first()
            return
        if index == self.size - 1:
            self.remove_last()
            return

        current = self.head
        for _ in range(index):
            current = current.next

        current.prev.next = current.next
        current.next.prev = current.prev
        self.size -= 1

    def get_index(self, index):
        if index < 0 or index >= self.size:
            raise IndexError('index out of range')

        current = self.head
        for _ in range(index):
            current = current.next
        return current.value

    def print_forward(self):
        current = self.head
        elements = []
        while current:
            elements.append(current.value)
            current = current.next
        print(" -> ".join(map(str, elements)))

    def print_backward(self):
        current = self.tail
        elements = []
        while current:
            elements.append(current.value)
            current = current.prev
        print(" <- ".join(map(str, elements)))
```

```python
dll = MyDoublyLinkedList()
dll.push_to_head(10)
dll.push_to_head(20)
dll.push_to_tail(5)
dll.push_to_index(1, 15) 

dll.print_forward()  
dll.print_backward()  

print("Размер списка:", dll.size)       
print("Элемент по индексу 2:", dll.get_index(2))  

dll.remove_to_index(1)
dll.print_forward() 
```
