```python
# В базе данных ich_edit три таблицы. Users с полями (id, name, age), Products с полями (pid, prod, quantity)
# и Sales с полями (sid, id, pid).
#
# Программа должна вывести все имена из таблицы users, дать пользователю выбрать одно из них и вывести все покупки
# этого пользователя.


import mysql.connector

def mysql_request(request_table_name, params = None):
    connection = None
    try:
        connection = mysql.connector.connect(host='ich-db.edu.itcareerhub.de',
                                     port=3306,
                                     user='ich1',
                                     password='password',
                                     db='ich_edit',)
        

        with connection.cursor(dictionary=True) as cursor:
            cursor.execute(request_table_name, params)
            result = cursor.fetchall()
            return result

    except mysql.connector.Error as ex:
        print(f"Database error [{ex.errno}]: {ex.msg}")
        return None

    finally:
        if connection and connection.is_connected():
            connection.close()


def get_user_users():
    request_users = 'select * from users;'
    res_users = mysql_request(request_users)
    return res_users
   

def get_user_product():
    id_user = int(input('Input user\'s ID: '))
    request_product = 'SELECT * FROM users u JOIN sales s ON u.id = s.id JOIN product p ON s.pid = p.pid WHERE u.id = %s;'
    res_product = mysql_request(request_product, params=(f"{id_user}",))
    return res_product
   

def print_table_users():
    res_users = get_user_users()
    if res_users:
        for row in res_users:
            print(row)


def print_table_product():
    res_product = get_user_product()
    if res_product:
        for prod in res_product:
            print(prod)
    else:
        print(f'No data found')
    
    
print_table_users()    
print_table_product()
```
