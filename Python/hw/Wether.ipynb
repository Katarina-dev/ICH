```python
import requests
#from bs4 import BeautifulSoup
from typing import Optional
import json

def get_response(url: str) -> Optional[requests.Response]:
    '''
    Функция отправляет GET-запрос по заданному URL-адресу и возвращает объект ответа.

    :param url: URL-адрес для отправки GET-запроса.
    :return: Объект ответа requests.Response или None в случае ошибки.
    '''
    try:
        response = requests.get(url)
        response.raise_for_status()        
        return response
    except requests.RequestException as e:
        print(f'Ошибка при запросе {url}: {e}')
        return None

def format_weather_data(weather_data):
    location = weather_data[0]["location"]
    current = weather_data[1]["current"]

    formatted_data = f'''
    Текущее состояние погоды в {location["name"]}:

    Температура: {current["temp_c"]}°C ({current["temp_f"]}°F)
    Состояние: {current["condition"]["text"]}
    Ветер: {current["wind_mph"]} миль/ч ({current["wind_kph"]} км/ч) из {current["wind_dir"]} направления ({current["wind_degree"]}°)
    Давление: {current["pressure_mb"]} мбар ({current["pressure_in"]} дюймов рт.ст.)
    Осадки: {current["precip_mm"]} мм ({current["precip_in"]} дюймов)
    Влажность: {current["humidity"]}%
    Облачность: {current["cloud"]}%
    Ощущаемая температура: {current["feelslike_c"]}°C ({current["feelslike_f"]}°F)
    Температура точки росы: {current["dewpoint_c"]}°C ({current["dewpoint_f"]}°F)
    Видимость: {current["vis_km"]} км ({current["vis_miles"]} миль)
    УФ индекс: {current["uv"]}
    Порывы ветра: {current["gust_mph"]} миль/ч ({current["gust_kph"]} км/ч)
    Местное время: {location["localtime"]}
    '''
    return formatted_data.strip()



url = 'http://api.weatherapi.com/v1/current.json?key=6e073286072a4e5cac7113931250601&q=London&aqi=no'

response = get_response(url)

if response is not None:
   
    print('Status Code:', response.status_code)
    print('-'*45)
    
    responce_text = response.text
    data = json.loads(responce_text)
   
    weather_data = [
        {"location": data["location"]}, 
        {"current": data["current"]}
    ]
    print(format_weather_data(weather_data))
    
    #for item in 
    #print(weather_data)
    
    #print('Response Text:', response.text)
    #print('-'*45)

else:
    print(f'Не удалось получить ответ от URL {url}')
```
