```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N = 10
x = np.linspace(-10, 10, N)
y = 3 * x**3 + 2 * x**2 + x + np.random.normal(0, 100, N)
display(x, y)
```

```python
k = np.polyfit(x, y, 3)

polynom = np.poly1d(k)

print("Коэффициенты полинома:")
print(f"a3 (x³) = {k[0]:.3f}")
print(f"a2 (x²) = {k[1]:.3f}")
print(f"a1 (x¹) = {k[2]:.3f}")
print(f"a0 (свободный член) = {k[3]:.3f}")

plt.scatter(x, y, label='Исходные данные')
plt.plot(x, polynom(x), label='Подогнанный полином')
plt.legend()
plt.title('Аппроксимация полиномом степени 3')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
display(titanic)

range_age = [0, 18, 35, 55, 85, 120]
name_age = ['child', 'young', 'adult', 'middle aged', 'aged']
titanic['age_group'] = pd.cut(titanic.age, bins = range_age, labels=name_age)
titanic.groupby(['age_group', 'pclass'])['survived'].mean().mul(100).round(2)
```
