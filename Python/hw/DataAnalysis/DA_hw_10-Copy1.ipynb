```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.petal_length.hist()	
```

```python
df1 = df[(df.petal_length >= 1) & (df.petal_length <= 2)]
df2 = df[(df.petal_length >= 3) & (df.petal_length <= 7)]

plt.figure(figsize=(9, 6))

sns.histplot(data=df1, x='petal_length', hue='species', binwidth=0.1, element="bars", shrink=0.5, color='darkblue', edgecolor=None, alpha=0.5, legend=False)

sns.histplot(data=df2, x='petal_length', hue='species', binwidth=0.1, element="step", palette=["#DD6600", "#218227"], edgecolor=None, alpha=0.5, legend=False)

plt.legend(title='Species', labels=['setosa', 'versicolor', 'virginica'])
plt.title("Iris Species Distribution by Petal Length")
plt.xlabel("Petal Length (cm)", fontsize=12)

plt.show()
```
