```python
Задача 5: Создайте возрастные группы и рассчитайте коэффициент выживаемости
для каждой возрастной группы в наборе данных Titanic.
```

```python
titanic = sns.load_dataset('titanic')
display(titanic)

range_age = [0, 18, 35, 55, 85, 120]
name_age = ['child', 'young', 'adult', 'middle aged', 'aged']
titanic['age_group'] = pd.cut(titanic.age, bins = range_age, labels=name_age)
titanic.groupby('age_group')['survived'].mean() * 100
```

```python
Задача 9: Заполните недостающие значения в столбце "Возраст" медианным
возрастом пассажиров того же класса и пола в наборе данных Titanic.
```

```python
titanic = sns.load_dataset('titanic')
display(titanic)

median_age = titanic.groupby(['class', 'sex']).age.median().reset_index()
df = titanic.merge(median_age, on = ['class', 'sex'])
df.age_x.isnull().sum()
df.age_y.isnull().sum()

df.age_x = df.age_x.fillna(df.age_y)
df.age_x.isnull().sum()
df.drop('age_y', axis = 1).rename(columns = {'age_x': 'age'})
```
