```python
plt.plot(range(10), [x**2 for x in range(10)], linestyle='--', color='green', marker='o')
plt.title("Квадратичная функция")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.show()
```

```python
plt.plot(range(10), [x**2 for x in range(10)], linestyle='--', color='green', marker='o', label='Парабола')
plt.plot(range(10), [x**3 for x in range(10)], linestyle='dotted', color='blue', marker='*', label='Кубическая парабола')
plt.title("Степенные функции")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid(True)
plt.legend()
plt.show()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=30, color='steelblue', edgecolor='black')
plt.title("Гистограмма распределения")
plt.xlabel("Значение")
plt.ylabel("Частота")
plt.show()
```

```python
data1 = np.random.randn(10000)
data2 = np.random.randn(2000)
plt.hist([data1, data2], bins=20)
plt.show()
plt.hist([data1, data2], bins=40, histtype='step', color=['red','yellow'])
plt.show()
```

```python
data = np.random.randint(0, 50, 100)
plt.hist(data, bins=10, color = 'red', edgecolor='green', range=(0,50), histtype='bar', cumulative=False, 
         orientation='vertical', alpha=1)
plt.show()
```

```python
data = np.random.randint(0,51,100)
plt.hist(data, bins=20, color='blue', edgecolor='black', range=(0,20), histtype = 'step', cumulative = True, 
         orientation = 'horizontal', alpha = 0.5)
plt.show()
```

```python
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0.2, 0) # выделяем второй сектор
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Круговая диаграмма")
plt.show()
```

```python
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
performance = [10,8,6,4,2,1]
plt.bar(objects, performance, color='blue', edgecolor='black')
plt.title("Популярность языков программирования")
plt.xlabel("Языки")
plt.ylabel("Популярность")
plt.show()
```

```python
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
performance2020 = [10,8,6,4,2,1]
performance2000 = [1,10,8,4,2,1]
index=np.arange(0,2*len(objects),2)
plt.bar(index, performance2020, color='blue', edgecolor='black', width=0.8)# от 0 до 0.8, от 2 до 2.8
plt.bar(index+0.8, performance2000, color='red', edgecolor='blue', width=0.8)# от 0.8 до 1.6, от 2.8 до 3.6
plt.xticks(index+0.4, objects)
plt.title("Популярность языков программирования")
plt.xlabel("Языки")
plt.ylabel("Популярность")
plt.show()
```

```python
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
performance2020 = [10,8,6,4,2,1]
performance2000 = [1,10,8,4,2,1]
index=np.arange(0,2*len(objects),2)
plt.bar(index, performance2020, color='blue', edgecolor='black', width=0.8, label='2020')# от 0 до 0.8, от 2 до 2.8
plt.bar(index+0.8, performance2000, color='red', edgecolor='blue', width=0.8, label='2000')# от 0.8 до 1.6, от 2.8 до 3.6
plt.xticks(index+0.4, objects)
plt.title("Популярность языков программирования")
plt.xlabel("Языки")
plt.ylabel("Популярность")
plt.legend()
plt.show()
```

```python
x = range(1, 5)
y = [1, 4, 6, 8]
plt.plot(x, y, marker='o')
plt.xlim(0, 5) # Установка пределов оси X
plt.ylim(0, 10) # Установка пределов оси Y
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.grid(True) # Включение отображения сетки
plt.show()
```

```python
plt.plot([1, 2, 3], [4, 3, 2])
plt.text(1, 4, "Start", fontsize=12)
plt.text(2, 3, "Middle", fontsize=12)
plt.text(3, 2, "End", fontsize=12)
plt.annotate('Important point', xy=(2, 3), xytext=(3, 4), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
x = np.arange(0, 10, 1)
y = x ** 2
plt.plot(x, y)
plt.title('Graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.text(2, 4, "Point")
plt.show()
```

```python
x = np.arange(0, 10, 1)
y = x ** 2
plt.plot(x, y, label='Plot')
plt.title('Graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.text(2, 4, "Point")
plt.bar(x, y, label='Bar')
plt.legend()
plt.show()
```
