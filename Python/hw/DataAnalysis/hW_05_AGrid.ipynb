1. Работа с временными рядами в pandas:


Напишите программу, которая:

- Создаёт DataFrame с двумя столбцами: "Дата" (в формате дд-мм-гггг) и "Значение" (целые числа).

- Преобразует столбец "Дата" в тип datetime и устанавливает его в качестве индекса DataFrame.

- Генерирует временной ряд, заполняя столбец "Значение" случайными числами.

- Выводит сгенерированный временной ряд.

- Запрашивает у пользователя начальную и конечную дату периода для фильтрации.

- Выводит временной ряд за ук

Пример вывода:


Сгенерированный временной ряд:

Дата

01-01-2023    120

02-01-2023    135

...

31-12-2023    110


Введите начальную дату (дд-мм-гггг): 01-01-2023

Введите конечную дату (дд-мм-гггг): 31-01-2023


Временной ряд за указанный период:

Дата

01-01-2023    120
.ит на экран данные о самом молодом участнике.

```python
import numpy as np
import pandas as pd

print('Определяем временной период')
start = '01-01-2025'
end = '28-02-2025'

start_date = pd.to_datetime(start, format='%d-%m-%Y')
end_date = pd.to_datetime(end, format='%d-%m-%Y')

daten = pd.date_range(start=start_date, end=end_date, freq='D')

values = np.random.randint(100, 200, size=len(daten))

df = pd.DataFrame({'Дата': daten, 'Значение': values})
df['Дата'] = df['Дата'].dt.strftime('%d-%m-%Y')
df.set_index('Дата', inplace=True)

print('Временной ряд:')
display(df)

try:
    start_input = input('Введите начальную дату в формате "дд-мм-гггг": ')
    end_input = input('Введите конечную дату в формате "дд-мм-гггг": ')
    
    start_input_date = pd.to_datetime(start_input, format='%d-%m-%Y').strftime('%d-%m-%Y')
    end_input_date = pd.to_datetime(end_input, format='%d-%m-%Y').strftime('%d-%m-%Y')
    
    filt_df = df.loc[start_input_date:end_input_date]
    
    print('\nВременной ряд за указанный период:')
    display(filt_df)
    
except ValueError:
    print('Ошибка! Введите ТОЛЬКО числа через дефис (дд-мм-гггг)')

df['Дата'] = pd.to_datetime(df['Дата'], format='%d-%m-%Y')
```

2. Работа с несколькими DataFrame


Напишите программу, которая:

- Создаёт DataFrame с тремя столбцами: "Товар", “Откуда”, “Куда” (в последних двух столбцах указываются города).

- Создаёт второй DataFrame с двумя столбцами: "Город", “Страна”.

- На основе слияния двух предыдущих DataFrame создает третий, в котором для каждого товара вместо городов доставки будут указаны страны. Лишних столбцов в этом DataFrame быть не должно, только "Товар", “Откуда”, “Куда” (в последних двух столбцах указываются страны)

- Добавьте в последний DataFrame новый столбец “Внутри_страны”, который принимает логические значения (True, если доставка осуществляется внутри страны).

- Вычислите, сколько в наших данных доставок внутри страны. 

```python
import numpy as np
import pandas as pd

countries = {
    "США": ["Нью-Йорк", "Лос-Анджелес", "Чикаго"],
    "Германия": ["Берлин", "Гамбург", "Мюнхен"],
    "Франция": ["Париж", "Марсель", "Лион"],
    "Великобритания": ["Лондон", "Бирмингем", "Манчестер"],
    "Китай": ["Пекин", "Шанхай", "Гуанчжоу"],
    "Индия": ["Дели", "Мумбаи", "Бангалор"],
    "Бразилия": ["Сан-Паулу", "Рио-де-Жанейро", "Бразилиа"],
    "Канада": ["Торонто", "Монреаль", "Ванкувер"],
    "Япония": ["Токио", "Осака", "Йокогама"]
}

box_post = {
    "Товар": ["Ноутбук","Смартфон","Наушники","Телевизор","Кофеварка","Микроволновка"],
    "Откуда": ["Нью-Йорк", "Гамбург", "Йокогама", "Марсель", "Гуанчжоу", "Шанхай"],
    "Куда": ["Чикаго", "Рио-де-Жанейро", "Токио", "Лион", "Торонто", "Париж"]
}

countries_df = pd.DataFrame({'Страна': list(countries.keys()), 'Город': list(countries.values())})
countries_df = countries_df.explode('Город')

box_post_df = pd.DataFrame(box_post)

display(countries_df.head())
box_post_df
```

```python
merge_out = pd.merge(box_post_df, countries_df, left_on='Откуда', right_on='Город', how='left') \
    .rename(columns={'Страна': 'Откуда (страна)'}).drop(columns=['Откуда', 'Город', 'Куда'])
display(merge_out)

merge_in = pd.merge(box_post_df, countries_df, left_on='Куда', right_on='Город', how='left') \
    .rename(columns={'Страна': 'Куда (страна)'}).drop(columns=['Откуда', 'Город', 'Куда'])
display(merge_in)
```

```python
merge_total = pd.merge(merge_out, merge_in, on=None, how='inner')
merge_total['Внутри страны'] = merge_total['Откуда (страна)'] == merge_total['Куда (страна)']
display(merge_total)
print(f"Доставок внутри страны {(merge_total['Внутри страны'] == True).sum()}")
```

3. Напишите программу, которая:

Запрашивает у пользователя количество строк данных для ввода.

Для каждой строки запрашивает данные по трем столбцам: "Имя", "Возраст", "Регион", разделённые пробелом.

Создает DataFrame на основе введенных данных.

Выводит общую информацию о DataFrame, включая количество записей, средний возраст, и количество уникальных регионов.

Выбирает и выводит на экран данные о самом молодом участнике.

```python
import numpy as np
import pandas as pd

try:
    n = int(input('Количество строк для ввода:'))
    
except ValueError:
    print('Ошибка! Введете ТОЛЬКО число без пробелов')
    exit()
    
print('Введите данные через пробел')
data = []

for i in range(n):
    row = input(f'{i + 1}. Имя Возраст Регион: ').split()
    if len(row) != 3:
        print('Ошибка! Введите ровно три значения через пробел: Имя Возраст Регион')
        continue
    name, age, region = row
    if not age.isdigit():
        print('Ошибка! Возраст должен быть числом.')
        continue
    data.append([name.title(), int(age), region.title()])

df = pd.DataFrame(data, columns=['Имя', 'Возраст', 'Регион'])

display(df)

print(f"""Количество записей: {n}
Средний возраст: {round(df['Возраст'].mean(), 1)} лет
Количество уникальных регионов: {df['Регион'].nunique()}""")

print(f"""Самый молодой участник:
{df.loc[df['Возраст'] == df['Возраст'].min()]}""")
```
