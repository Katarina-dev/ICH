```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
df.age = df.age.fillna(df.age.mean())
df
```

```python
import random

def get_data_for_deck(row):
    if pd.isnull(row.deck):
        if row.pclass in [1, 2]:
            return random.choice(['B', 'C', 'D'])
        elif row.pclass == 3:
            return random.choice(['E', 'F', 'G'])
    return row.deck
            
df.deck = df.apply(get_data_for_deck, axis=1)
df
```

```python
df.info()
```

```python
print(df.embark_town.unique())
```

```python
df.embark_town = df.embark_town.fillna(df.embark_town.mode()[0])
df
```

```python
print(df.embark_town.unique())
```

```python
print(df.embarked.unique())
```

```python
def get_data_for_embarked(row):
    if pd.isnull(row.embarked):
        if row.embark_town == 'Southampton':
            return 'S'
        elif row.embark_town == 'Cherbourg':
            return 'C'
        elif row.embark_town == 'Queenstown':
            return 'Q'
    return row.embarked
            
df.embarked = df.apply(get_data_for_embarked, axis=1)
df
```
