```python
Задача: Создайте простое приложение Dash, которое позволяет пользователям выбрать континент из выпадающего меню и динамически отображает гистограмму ожидаемой продолжительности жизни для выбранного континента, используя набор данных Gapminder.

Инструкции:

Инициализируйте приложение Dash.

Определите макет, включающий dcc.Dropdown для выбора континента и dcc.Graph для отображения гистограммы.

Наполните выпадающие опции континентами из набора данных Gapminder.

Напишите функцию обратного вызова, которая обновляет гистограмму в зависимости от выбранного континента.

Гистограмма должна отображать распределение lifeExp для выбранного континента.

Добавьте к гистограмме соответствующие заголовки и метки осей.
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from dash import Dash, html, dcc
import plotly.express as px
from dash import Input, Output, callback
```

```python
app = Dash(__name__)

df = px.data.gapminder()

header = html.Header(html.H1('Dashboard \'Continents\'', className='header-title'))

dropdown_component = dcc.Dropdown(
    id='my_dropdown',
    options=[
        {'label': 'All', 'value': ''},
        {'label': 'Asia', 'value': 'Asia'},
        {'label': 'Africa', 'value': 'Africa'},
        {'label': 'Americas', 'value': 'Americas'},
        {'label': 'Europe', 'value': 'Europe'},
        {'label': 'Oceania', 'value': 'Oceania'}
    ],
    value='',
    placeholder='Select continent'
)


sidebar = html.Div([
    html.H2('Navigation', className='sidebar-title'),
    dropdown_component 
], className='sidebar')

graph_component = dcc.Graph(id='example-graph')

app.layout = html.Div([
    header,
    sidebar,
    graph_component
])

@callback(
    Output('example-graph', 'figure'),
    Input('my_dropdown', 'value')
    
)
def update_graph(selected_continent):
    if selected_continent:  
        filtered_df = df[df['continent'] == selected_continent]
    else:  
        filtered_df = df  
    diag = px.histogram(filtered_df, x='lifeExp', nbins=20, title = 'Life expendancy by continents')

    diag.update_layout(
        xaxis_title = 'Life expendancy',
        yaxis_title = 'Frequency',
        bargap=0.2
    )

    return diag
    

if __name__ == '__main__':
    app.run(debug=True)
```
