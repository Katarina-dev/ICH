# Меры дисперсии

```python
Меры дисперсии - это статистические инструменты, используемые для
описания разброса или изменчивости в наборе данных.
Они позволяют понять, насколько сильно точки данных отличаются от центрального
значения (среднего, медианы или моды). Три наиболее распространенные меры
дисперсии - это диапазон, дисперсия и стандартное отклонение.
```

## Диапазон

```python
Диапазон - это самая простая мера дисперсии. Он рассчитывается как разница
между максимальным и минимальным значениями в наборе данных. Диапазон дает
быстрое представление о разбросе данных, но очень чувствителен к выбросам.
```

```python
data = [4, 8, 6, 5, 3, 2, 8, 9, 2, 5]
range_value = max(data) - min(data)
print("Диапазон:", range_value)
```

## Вариация

```python
Вариация (дисперсия) - это среднее квадратичное отклонение каждой точки данных от среднего
значения. Это количественная оценка разброса точек данных. Более высокая
дисперсия указывает на то, что точки данных больше удалены от среднего значения,
в то время как более низкая дисперсия указывает на то, что они ближе к среднему
значению.
```

```python
import numpy as np
data = [4, 8, 6, 5, 3, 2, 8, 9, 2, 5]
variance = np.var(data)
print("Вариация:", variance)
```

## Стандартное отклонение

```python
Стандартное отклонение - это квадратный корень из дисперсии. Оно представляет
собой меру среднего расстояния каждой точки данных от среднего значения.
Стандартное отклонение широко используется, поскольку оно выражается в тех же
единицах, что и данные, что облегчает его интерпретацию
```

```python
import numpy as np
data = [4, 8, 6, 5, 3, 2, 8, 9, 2, 5]
std_deviation = np.std(data)
print("Стандартное отклонение:", std_deviation)
```

```python
Практический пример: Вычисление диапазона, дисперсии и стандартного
отклонения с помощью Python
Рассмотрим набор данных и рассчитаем диапазон, дисперсию и стандартное
отклонение с помощью библиотеки NumPy в Python.
```

```python
data = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

range_value = max(data) - min(data)
print("Диапазон:", range_value)

variance = np.var(data)
print("Вариация:", variance)

std_deviation = np.std(data)
print("Стандартное отклонение:", std_deviation)
```

```python
● Диапазон: Разница между максимальным и минимальным значениями в
наборе данных. Позволяет быстро оценить разброс, но чувствителен к
выбросам.

● Дисперсия: Среднее квадратичное отклонение от среднего значения.
Количественно характеризует общий разброс точек данных.
    
● Стандартное отклонение: Квадратный корень из дисперсии. Оно позволяет
оценить среднее расстояние каждой точки данных от среднего значения и
легче интерпретируется, поскольку находится в тех же единицах, что и
данные.
```

```python
Проверка гипотез - это статистический метод, используемый для того, чтобы
сделать выводы о совокупности на основе выборочных данных.
```

```python
Он предполагает формулирование двух конкурирующих гипотез: нулевой и
альтернативной. Цель - определить, достаточно ли доказательств в данных выборки,
чтобы отвергнуть нулевую гипотезу в пользу альтернативной гипотезы.
```

```python
Нулевая гипотеза (H0):
Нулевая гипотеза - это утверждение об отсутствии эффекта или различий,
представляющее собой статус-кво или базовое предположение.

Пример: Если мы проверяем, является ли монета честной, нулевая гипотеза может
состоять в том, что вероятность выпадения решек равна 0,5 .

Альтернативная гипотеза (H1 или Ha):
Альтернативная гипотеза - это утверждение, противоречащее нулевой гипотезе. Она
представляет собой эффект или разницу, которую исследователь стремится
обнаружить.

Пример: Для теста на честность монеты альтернативная гипотеза может
заключаться в том, что вероятность выпадения решек не равна 0,5.
```

```python
Шаги при проверке гипотез:

Сформулируйте гипотезы:
Сформулируйте нулевую и альтернативную гипотезы.
    
Выберите уровень значимости (α):
Уровень значимости, часто обозначаемый альфа (α), - это вероятность отвергнуть
нулевую гипотезу, если она верна. Обычно выбирают 0,05, 0,01 и 0,10.

Выберите подходящий тест:
Выберите статистический тест в зависимости от типа данных и гипотезы. К
распространенным тестам относятся t-тесты, тесты хи-квадрат, ANOVA и
регрессионный анализ.

Соберите данные:
Соберите данные, которые будут анализироваться в тесте.
    
Рассчитайте статистику теста:
На основе собранных данных и выбранного теста рассчитайте тестовую статистику,
которая отражает, насколько наблюдаемые данные отклоняются от нулевой
гипотезы.
    
Определите p-значение:
p-значение - это вероятность того, что результаты теста будут не менее
экстремальными, чем наблюдаемые результаты, при условии, что нулевая гипотеза
верна.

Примите решение:
Сравните p-значение с выбранным уровнем значимости:
Если p-значение ≤ α: Отклоните нулевую гипотезу.
Если p-значение > α: Не отвергайте нулевую гипотезу.
    
Представьте результаты:
Представьте результаты проверки гипотез, включая тестовую статистику,
p-значение и вывод о гипотезах.
    
Пример t-теста
Т-тест используется для сравнения средних значений двух групп. Здесь мы
проведем независимый двухвыборочный t-тест с помощью библиотеки SciPy в
Python.
```

```python
from scipy import stats
# Данные выборки
data1 = [1, 2, 3, 4, 5]
data2 = [6, 7, 8, 9, 10]
# Выполните t-тест
t_statistic, p_value = stats.ttest_ind(data1, data2)
print('t-статистика:', t_statistic)
print('p-value:', p_value)
```

```python
В этом примере p-значение меньше 0,05, поэтому мы отвергаем нулевую гипотезу и
делаем вывод о наличии значимой разницы между средними значениями двух
групп.
```

```python
Пример теста хи-квадрат
Тест хи-квадрат используется для определения наличия значимой связи между
двумя категориальными переменными. Здесь мы проведем тест хи-квадрат с
помощью библиотеки SciPy в Python
```

```python
from scipy.stats import chi2_contingency
import numpy as np
# Таблица контингентности
data = np.array([[207, 282, 241], [234, 242, 232]])
# Выполните тест хи-квадрат
chi2, p, dof, expected = chi2_contingency(data) # хи-квадрат всегд принимает за нулевую гипотезу, что переменные не зависят друг от друга
print('Статистика хи-квадрат:', chi2)
print('p-значение:', p)
print('Степени свободы:', dof)
print('Ожидаемые частоты:', expected)
```

```python
В этом примере p-значение больше 0,05, поэтому мы не отвергаем нулевую гипотезу
и делаем вывод об отсутствии значимой связи между двумя категориальными
переменными.
```

# Доверительные интервалы

```python
Доверительный интервал (ДИ) - это диапазон значений, полученный на
основе выборочных данных, который, скорее всего, содержит истинное значение
неизвестного параметра генеральной совокупности.
```

```python
Интервал имеет соответствующий уровень доверия, который определяет степень
уверенности в том, что параметр находится в пределах интервала. Обычно
используются уровни доверия 90%, 95% и 99%.
Ключевые понятия
1. Уровень доверия: Вероятность того, что доверительный интервал содержит
истинный параметр генеральной совокупности. Например, уровень доверия
95 % означает, что если мы возьмем 100 различных выборок и рассчитаем
доверительный интервал для каждой выборки, мы будем ожидать, что около
95 интервалов будут содержать параметр генеральной совокупности.
2. Предел погрешности: Диапазон выше и ниже статистики выборки в
доверительном интервале. На него влияют размер выборки, изменчивость
данных и уровень доверия.
```

```python
Пример: Вычисление доверительных интервалов с помощью Python
Давайте рассчитаем 95% доверительный интервал для выборочного набора данных
с помощью библиотек scipy и numpy в Python.
```

```python
import numpy as np
from scipy import stats

#Определите выборочные данные:
data = [99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86]

#Вычислите выборочное среднее и стандартное отклонение:
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1) # ddof=1 для выборочного стандартного отклонения
n = len(data)

#Определите Z-значение для 95-процентного уровня доверия:
confidence_level = 0.95
alpha = 1 - confidence_level
z_value = stats.norm.ppf(1 - alpha/2)

#Вычислите предел ошибки:
margin_of_error = z_value * (sample_std / np.sqrt(n))

#Рассчитайте доверительный интервал:
confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)
print("95% доверительный интервал:", confidence_interval)
```

```python
Данная выборка говорить нам о том, что среднее значение скорее всего лежит в интервале (np.float64(84.53095502435286), np.float64(95.00750651410868))

Если бы мы взяли много выборок и рассчитали доверительный интервал для каждой, то около 95 %
этих интервалов содержали бы истинное среднее значение генеральной совокупности.
```

```python
1. Определите, значительно ли отличается средний рост выборки студентов от
известного среднего роста генеральной совокупности.
Известно, что средний рост студентов в определенной генеральной
совокупности составляет 170 см. У вас есть выборка из 10 студентов со
следующим ростом (в см): [172, 168, 171, 169, 170, 173, 167, 174, 175, 169].
```

```python
import numpy as np
from scipy import stats

sample_heights = [172, 168, 171, 169, 170, 173, 167, 174, 175, 169]
population_mean = 170
# H0 - средний рост не отличается от 170 см
t_statistic, p_value = stats.ttest_1samp(sample_heights, population_mean)

print('t-statistic:', t_statistic)
print('p-value:', p_value)
print("H0 - средний рост равен 170 см")
if p_value > 0.05:
    print('гипотезу не отвергаем')
else:
    print('гипотезу  отвергаем')
np.mean(sample_heights)
```

```python
import numpy as np
from scipy import stats

class_a_scores = [85, 88, 90, 92, 87, 85, 89, 91, 90, 86]
class_b_scores = [78, 82, 84, 79, 81, 80, 83, 85, 84, 82]

# H0 - нет значимой разницы 
t_statistic, p_value = stats.ttest_ind(class_a_scores, class_b_scores)
print(np.mean(class_a_scores))
print(np.mean(class_b_scores))

print('t-statistic:', t_statistic)
print('p-value:', p_value)
```

```python
3. Определите, существует ли значимая связь между полом и предпочтением
нового продукта.
Был проведен опрос с целью выяснить, существует ли связь между полом и
предпочтениями в отношении нового продукта. Результаты оказались
следующими:
Мужчины, которым нравится продукт: 30
Мужчины, которым продукт не нравится: 20
Женщины, которым нравится продукт: 25
Женщины, которым не нравится продукт: 25
```

```python
import numpy as np
from scipy.stats import chi2_contingency

# H0 мужчины и женищины  ведут себя независимо ставя лайки и дислайки
data = np.array([[50, 5], [25, 25]])

chi2, p, dof, expected = chi2_contingency(data) 

print('Chi-square statistic:', chi2)
print('p-value:', p)
print('Degrees of freedom:', dof)
print('Expected frequencies:', expected)

print('\nH0 - мужчины и женищины  ведут себя независимо ставя лайки и дислайки')
if p > 0.05:
    print("гипотезу оставляем")
else:
    print("гипотезу отвергаем")
```
