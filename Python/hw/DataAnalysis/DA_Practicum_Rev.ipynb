```python
Загрузите датасет  "Пингвины" из seaborn и познакомьтесь с ним:

Постройте диаграмму рассеяния по значениям bill_length_mm и bill_depth_mm. 
Добавьте цветовое выделение групп по столбцу sex. 
Постройте круговую диаграмму, показывающую распределение пингвинов датасета по островам.
Постройте столбчатую диаграмму, которая показывает среднюю массу пингвинов для каждого вида и пола.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")

print(penguins.head())

# ● Постройте диаграмму рассеяния по значениям bill_length_mm и bill_depth_mm. Добавьте цветовое выделение групп по столбцу sex.
sns.scatterplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='sex') 
plt.title('Scatter Plot') 
plt.show()

# ● Постройте круговую диаграмму, показывающую распределение пингвинов датасета по островам. 
penguins.groupby('island').size().plot.pie()

# ● Постройте столбчатую диаграмму, которая показывает среднюю массу пингвинов для каждого вида и пола.
penguins.groupby(['species', 'sex']).body_mass_g.mean().unstack().plot(kind='bar')
```

```python
Задача 4: Создайте линейный график набора данных flights, показывающий
количество пассажиров с течением времени. Пометьте график максимальными и
минимальными значениями.
```

```python
flights = sns.load_dataset('flights')
display(flights)
```

```python
sns.lineplot(x='year', y='passengers', hue = 'month', data=flights)
plt.title('Линейный график')
plt.show()
```

```python
dic = {'Jan': 1,
       'Feb': 2,
       'Mar': 3,
       'Apr': 4,
       'May': 5,
       'Jun': 6,
       'Jul': 7,
       'Aug': 8,
       'Sep': 9,
       'Oct': 10,
       'Nov': 11,
       'Dec': 12}
flights['month_1'] = flights['month'].map(dic).fillna(1).astype(int)
flights
       
```

```python
flights['date'] = flights['year'] + flights['month_1']/12
flights
```

```python
max_p = flights['passengers'].max()

max_date = flights[flights['passengers'] == max_p]['date']
print(max_date)

sns.lineplot(x='date', y='passengers', data=flights)
plt.annotate('max', xy = (max_date, max_p), xytext = (max_date-5, max_p), arrowprops=dict(facecolor = 'black', shrink=0.01))
plt.title('Линейный график')
plt.show()
```

```python
Задача 5: Создайте совместный график набора данных mpg, показывающий связь
между лошадиными силами и расходом топлива с помощью линии регрессии и KDE.
```
