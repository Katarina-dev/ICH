# Основы построения графиков с помощью Matplotlib

```python
Matplotlib - это универсальная библиотека для создания широкого спектра
статических, анимированных и интерактивных графиков на Python.
Ниже мы рассмотрим, как создавать основные линейные графики, графики
рассеяния и гистограммы с помощью Matplotlib.
```

## Линейные графики

```python

Линейные графики - это один из самых распространенных типов графиков,
используемых для визуализации тенденций изменения данных за интервалы
времени. Чтобы создать базовый линейный график в Matplotlib, вы можете
использовать функцию plot из подмодуля pyplot.
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Создаем квадратичную кривую
x = np.linspace(-10, 10, 100)
y = 3.5 - 2.3 * x + 0.5 * x ** 2
# Строим график данных
plt.plot(x, y)
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Основной линейный график')
plt.show()
```

## Графики рассеяния

```python
Графики рассеяния используются для отображения значений, как правило, двух
переменных для набора данных. Функция scatter в pyplot используется для создания
диаграмм рассеяния.
```

```python
# Генерируем случайные данные
x = np.random.randn(200)
y = np.random.randn(200)
# Создаём диаграмму рассеяния
plt.scatter(x, y, color='blue', marker='o')
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Основной график рассеяния')
plt.show()
```

## Столбчатые диаграммы

```python
Гистограммы полезны для сравнения величин, относящихся к разным категориям.
Для создания гистограмм используется функция bar в pyplot
```

```python
# Определяем категории и значения
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 2]
# Создаём гистограмму
plt.bar(categories, values, color='green')
plt.xlabel('Категории')
plt.ylabel('Значения')
plt.title('Основная гистограмма')
plt.show()
```

## Гистограммы

```python
Гистограммы используются для представления распределения набора данных. Для
создания гистограмм используется функция hist в pyplot.
```

```python
# Генерируем случайные данные
data = np.random.randn(1000)
# Создаём гистограмму
plt.hist(data, bins=30, color='purple')
plt.xlabel('Значение')
plt.ylabel('Частота')
plt.title('Основная гистограмма')
plt.show()
```

## Подграфики

```python
Подграфики - это функция в Matplotlib, позволяющая создавать фигуры,
содержащие несколько графиков, которые могут быть расположены в различных
макетах.
Это особенно полезно для сравнения различных наборов данных или визуализации
различных аспектов одного и того же набора данных.
```

```python
Функция plt.subplots() - это удобный способ создать фигуру и сетку подграфиков за
один вызов. Она возвращает объект фигуры и массив объектов осей.
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Создаем примерные данные
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
# Создайте рисунок с двумя подплощадками
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
# Нанесите данные на первый подграфик
ax1.plot(x, y)
ax1.set_title('Line Plot')
# Постройте данные на втором подграфике
ax2.scatter(x, y)
ax2.set_title('Scatter Plot')
# Отобразите рисунок
plt.show()
```

## Расширенные конфигурации подплотов

```python
Для более сложных конфигураций Matplotlib предоставляет такие инструменты, как
GridSpec и subplot_mosaic.
GridSpec: Позволяет гибко позиционировать подграфики в сетке. Подграфики могут
охватывать несколько строк или столбцов.
```

```python
from matplotlib.gridspec import GridSpec
fig = plt.figure(figsize=(10, 8))
gs = GridSpec(3, 3, figure=fig)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, :-1])
ax3 = fig.add_subplot(gs[1:, -1])
ax4 = fig.add_subplot(gs[-1, 0])
ax5 = fig.add_subplot(gs[-1, -2])
# Построение данных на подграфиках
ax1.plot(x, y)
ax2.plot(x, y)
ax3.plot(x, y)
ax4.plot(x, y)
ax5.plot(x, y)
plt.show()
```

```python
subplot_mosaic: новый метод, позволяющий задавать структуру фигуры с помощью
форматированных строк.
```

```python
layout = """
AB
CC
"""
fig, axes = plt.subplot_mosaic(layout, figsize=(8, 6))
axes['A'].plot(x, y)
axes['B'].scatter(x, y)
axes['C'].bar(np.arange(10), np.random.rand(10))
plt.show()
```

```python
layout = """
ABB
CCC
"""
fig, axes = plt.subplot_mosaic(layout, figsize=(8, 6))
axes['A'].plot(x, y)
axes['B'].scatter(x, y)
axes['C'].bar(np.arange(10), np.random.rand(10))
plt.show()
```

# Основы построения графиков с Seaborn

## Графики распределения

```python
Графики распределения используются для визуализации распределения набора
данных. Сиборн предлагает несколько функций для создания таких графиков,
включая гистограммы, оценки плотности ядра (KDE) и rug графики.
Гистограммы и графики KDE: Функция displot может быть использована для
создания гистограмм и графиков KDE. (Функция kdeplot предназначена специально
для построения графиков KDE.)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Загружаем данные выборки
data = sns.load_dataset("iris")
# Гистограмма с помощью KDE
sns.displot(data['sepal_length'], kde=True)
plt.title('Гистограмма с KDE')
plt.show()
```

## Rug графики

```python
Rug графики: Rug графики добавляют небольшие вертикальные линии в каждой
точке данных вдоль оси x, обеспечивая визуальное представление распределения
данных.
```

```python
sns.kdeplot(data['sepal_length'], fill=True)
sns.rugplot(data['sepal_length'])
plt.title('KDE Plot with Rug Plot')
plt.show()
```

## Категориальные графики

```python
Категориальные графики используются для визуализации данных, в которых одна
или несколько переменных являются категориальными. Seaborn предоставляет
несколько функций для создания категориальных графиков, включая столбчатые,
ящик с усами и скрипичные графики.
Столбчатые графики: Функция barplot используется для создания столбчатых
диаграмм, которые показывают связь между категориальной и непрерывной
переменными.
```

```python
# Гистограмма
sns.barplot(x='species', y='sepal_length', data=data)
plt.title('Bar Plot')
plt.show()
```

## Ящик с усами

```python
Функция boxplot создает графики ящик с усами, которые
отображают распределение непрерывной переменной для каждой категории
категориальной переменной.
```

```python
sns.boxplot(x='species', y='sepal_length', data=data)
plt.title('Box Plot')
plt.show()
```

## Скрипичные графики

```python
Функция violinplot сочетает в себе аспекты box plot и KDE plots,
показывая распределение данных по различным категориям.
```

```python
sns.violinplot(x='species', y='sepal_length', data=data)
plt.title('Скрипичный график')
plt.show()
```

## Реляционные графики

```python
Реляционные графики используются для визуализации взаимосвязи между двумя
или более переменными. Seaborn предоставляет такие функции, как relplot,
scatterplot и lineplot для создания этих графиков.
```

## Диаграмма рассеяния

```python
Функция scatterplot используется для создания диаграмм
рассеяния, которые показывают связь между двумя переменными.
```

```python
# Диаграмма рассеяния
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Scatter Plot')
plt.show()
```

## Линейные графики

```python
 Функция lineplot используется для создания линейных графиков,
которые полезны для визуализации тенденций с течением времени или другой
непрерывной переменной.
```

```python
# Линейный график
sns.lineplot(x='sepal_length', y='sepal_width', data=data)
plt.title('Линейный график')
plt.show()
```

## Реляционные графики с помощью relplot

```python
Функция relplot - это высокоуровневый
интерфейс для создания графиков разброса и линий. Ее можно настроить с
помощью параметра kind.
```

```python
# Реляционный график (разброс)
sns.relplot(x='sepal_length', y='sepal_width', data=data, kind='scatter')
plt.title('Реляционный график (разброс)')
plt.show()
```

```python
# Реляционный график (линия)
sns.relplot(x='sepal_length', y='sepal_width', data=data, kind='line')
plt.title('Реляционный график (линия)')
plt.show()
```

```python
Реляционные графики также могут включать дополнительные измерения с помощью
параметров hue, size и style для представления большего количества переменных
```

```python
# Реляционный график с оттенком и стилем
sns.relplot(x='sepal_length', y='sepal_width', hue='species', style='species',
data=data)
plt.title('Реляционный график с оттенком и стилем')
plt.show()
```

# Парные графики и тепловые карты в Seaborn

## Парные графики

```python
Парные графики используются для визуализации парных отношений в наборе
данных. Они особенно полезны для изучения взаимосвязей между несколькими
переменными одновременно. Функция pairplot в Seaborn создает сетку осей таким
образом, что каждая числовая переменная в наборе данных распределяется по
осям y в одной строке, а по осям x - в одном столбце. Диагональные графики
показывают одномерное распределение данных в каждом столбце.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Загрузка данных выборки
penguins = sns.load_dataset("penguins")
# Создайте парный график
sns.pairplot(penguins)
plt.show()
```

```python
# Парный график с оттенком
sns.pairplot(penguins, hue="species")
plt.show()
# Парный график с различными видами графиков
sns.pairplot(penguins, kind="kde")
plt.show()
# Парный график с пользовательскими маркерами и размерами
sns.pairplot(penguins, hue="species", markers=["o", "s", "D"], height=2.5)
plt.show()
```

```python
● hue: Добавляет семантическое отображение для окрашивания точек по
категориальной переменной.
● kind: Определяет тип используемого графика (например, 'scatter', 'kde', 'hist').
● markers (маркеры): Указывает разные маркеры для разных уровней
переменной оттенка.
● height (высота): Управляет размером каждой подплощадки.
```

## Тепловые карты

```python
Тепловые карты - это популярная техника визуализации данных, которая использует
цвет для представления различных уровней величины данных. Они особенно
полезны для визуализации корреляционной матрицы между несколькими
переменными, позволяя с первого взгляда определить высококоррелированные или
обратнокоррелированные переменные.
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Загрузка данных выборки
data = sns.load_dataset("iris")
data = data[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
# Рассчитайте корреляционную матрицу
correlation_matrix = data.corr()
# Создайте тепловую карту
sns.heatmap(correlation_matrix, cmap='coolwarm')
plt.show()
```
