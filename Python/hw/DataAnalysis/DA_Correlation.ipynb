```python
Тест Хи-квадрат - это статистический метод, широко используемый при
проверке гипотез, в частности при анализе категориальных данных.
```

```python
Тест Хи-квадрат на добросовестность соответствия. 
Условие: вам сказали проценты сортов ирисов в поле
Вы набрали ирисов и хотите узнать правду ли вам сказали
0-гипотеза: правда

Тест независимости (или ассоциации) Хи-квадрат
Условие: есть цвет глаз и высшее образование
Вы хотите узнать зависят ли они друг от друга?
0-гипотеза: не зависит
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
Нулевая гипотеза: продукты лайкаются одинаково
```

```python
Создайте или загрузите таблицу контингентности
Таблица случайностей - это матрица, которая отображает распределение частот
переменных. Для этого примера давайте создадим простую таблицу случайностей
вручную. В качестве альтернативы можно загрузить набор данных и создать из него
таблицу случайностей.
```

```python
# Пример таблицы случайностей
data = {'Preference': ['Like', 'Dislike', 'Neutral'],
'Product_A': [20, 15, 25],
'Product_B': [30, 10, 20]}
df = pd.DataFrame(data)
df.set_index('Preference', inplace=True)
print(df)
```

```python
Выполните тест хи-квадрат
Используйте функцию chi2_contingency из scipy.stats, чтобы выполнить тест
Хи-квадрат для таблицы случайностей.
```

```python
chi2, p, dof, expected = chi2_contingency(df)
print(f"Статистика хи-квадрат: {chi2}")
print(f"P-value: {p}")
print(f"Степени свободы: {dof}")
print("Ожидаемые частоты:")
print(expected)
```

```python
1. Рассчитайте коэффициент корреляции Пирсона между длиной и шириной
лепестка в наборе данных Iris.
```

```python
import pandas as pd
from scipy.stats import pearsonr
import seaborn as sns

iris = sns.load_dataset('iris')

# iris = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
iris.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width','class']
corr, _ = pearsonr(iris['petal_length'], iris['petal_width'])
print(f'Pearson correlation: {corr}')
```

```python
Если statistic(коэф корреляции) около нуля - нет корреляции
Если ближе к 1 или к -1, то корреляция есть
```

```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
from scipy.stats import chi2_contingency

# Загрузите набор данных
data = load_breast_cancer(as_frame=True)
df = data.frame
display(df)
```

```python
df['size_can'] = df['mean area'].apply(lambda x: 'big' if x > df['mean area'].median() else 'small')
```

```python
df['type_can'] = df['target'].apply(lambda x: 'zlo' if x == 0 else 'dobro')
```

```python
df2 = pd.crosstab(df['size_can'], df['type_can'])
df2
```

```python
chi2, p, dof, expected = chi2_contingency(df2)
print(f"Статистика хи-квадрат: {chi2}")
print(f"P-value: {p}")
print(f"Степени свободы: {dof}")
print("Ожидаемые частоты:")
print(expected)
```
