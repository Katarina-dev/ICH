```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
df.petal_length.hist()	
```

```python
df.petal_length.dtype
```

```python
print(df[(df.petal_length == 5)])
```

```python
print(df[(df.petal_length == 5)].shape[0])
```

```python
df
```

```python
sns.histplot(data=df, x='petal_length', hue='species', binwidth=0.1, element="bars", shrink=0.5)
```

```python
pv_data = df.pivot_table(index='species', columns='petal_length', aggfunc='size', fill_value=0)
pv_data
```

```python
df1 = df[(df.petal_length >= 1) & (df.petal_length <= 2)]
df2 = df[(df.petal_length >= 3) & (df.petal_length <= 7)]

plt.figure(figsize=(9, 6))

sns.histplot(data=df1, x='petal_length', hue='species', binwidth=0.1, element="bars", shrink=0.5, color='darkblue', edgecolor=None, alpha=0.5, legend=False)

sns.histplot(data=df2, x='petal_length', hue='species', binwidth=0.1, element="step", palette=["#DD6600", "#218227"], edgecolor=None, alpha=0.5, legend=False)

plt.legend(title='Species', labels=['setosa', 'versicolor', 'virginica'])

plt.show()
```

```python
pv3_data = df.groupby(['petal_length', 'species']).size().reset_index(name='count')
pv3_data
```

```python
print(pv3_data[(pv3_data.petal_length == 5)].shape[0])
```

```python
sns.histplot(data=pv3_data, x='petal_length', hue='species', weights='count', binwidth=0.1, element="step", shrink=0.5)
```
