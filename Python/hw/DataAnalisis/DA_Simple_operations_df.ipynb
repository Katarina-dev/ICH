```python
Врач на протяжении дня измеряет пациенту температуру каждые 3 часа в течение 2
недель. Также пациенту необходимо спать с 11 вечера до 7 утра. Каждый день
измерения температуры начинаются в 8 часов. Первое измерение 22 марта 2020
года. Определите моменты времени, когда нужно измерить пациенту температуру.
```

```python
start_time = '2020-03-22 08:00:00' 
time_range = pd.date_range(start = start_time, periods = 14 * 8-2, freq = '3h')
time_range
```

```python
filter_time = time_range[(time_range.hour >= 7) & (time_range.hour < 23)]
filter_time
```
