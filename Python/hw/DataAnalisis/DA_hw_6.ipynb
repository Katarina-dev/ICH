```python
import matplotlib.pyplot as plt

np.random.seed(42)
labels = [str(i) for i in range(1, 26)]
sizes = np.random.randint(100, 1000, 25)
plt.figure(figsize=(9, 9))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, counterclock=False)
plt.title("Энергопотребление каждой квартиры")
plt.show()
```

```python
objects = [str(i) for i in range(1, 26)]
performance = np.random.randint(100, 1000, 25)
plt.bar(objects, performance, color='violet', edgecolor='black', width=0.2)
plt.title("Энергопотребление каждой квартиры")
plt.xlabel("Квартира")
plt.ylabel("Энергопотребление")
plt.show()
```
