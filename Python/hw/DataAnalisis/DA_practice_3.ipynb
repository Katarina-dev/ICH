```python
Пять спортсменов участвовали в соревновании, в котором нужно было бежать 2 часа, плыть 1 час и ехать на велосипеде 5 часов.
Первый спортсмен бежал со скоростью 15 км/ч, плыл со скоростью 4 км/ч и ехал на велосипеде со скоростью 25 км/ч.
Соответствующие скорости второго спортсмена: 14 км/ч, 5 км/ч, 27 км/ч.
Третий спортсмен: 16 км/ч, 5 км/ч, 23 км/ч.
Четвертый: 13 км/ч, 5.5 км/ч, 26 км/ч.
Пятый: 12 км/ч, 4.5 км/ч, 30 км/ч.
Побеждает тот, кто преодолел большее расстояние. Напишите программу, которая определяет победителя.
(Подсказка: Подумайте, что здесь можно записать матрицей, что вектором, какая матричная операция определяет пройденное расстояние.)
```

```python
speeds = np.array([[15, 4, 25], 
                   [14, 5, 27],
                   [16, 5, 23],
                   [13, 5.5, 26],
                   [12, 4.5, 30]])
time = np.array([[2], [1], [5]])
print(speeds @ time)
distance = speeds @ time
np.argmax(distance) + 1
```

```python
Сделайте срез матрицы из предыдущей задачи, чтобы получить квадратную матрицу (т.е. оставляем только трех спортсменов). 
Сгенерируйте случайный столбец трех чисел от 100 до 200. Будем считать, что это расстояния, пройденные каждым спортсменом. 
Найдите определитель полученной матрицы и матрицу, обратную к ней. 
Умножьте обратную матрицу на случайный столбец (с помощью матричного умножения). Поясните, что вы нашли с помощью такой операции.
```

```python
sqw_matrix = speeds[0:3]
print(sqw_matrix)

arr = np.random.randint(100, 103, 3).reshape(3, 1)
print(arr)

det = np.linalg.det(sqw_matrix)
print(det)

if det != 0:
    inv_matrix = np.linalg.inv(sqw_matrix)
else:
    print('Matrix is not det')
print(f'Inverter Matrix: {inv_matrix}')

print(inv_matrix @ arr)
```
