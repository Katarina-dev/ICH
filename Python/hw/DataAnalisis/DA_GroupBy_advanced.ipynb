```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)
df = pd.DataFrame({
'Специальность' : ['Ветеринар', 'Ветеринар',
'Психолог', 'Психолог'] * 6,
'Врач' : ['Андрей', 'Сергей', 'Ирина'] * 8,
'Диагноз' : ['Простуда', 'Простуда', 'Простуда',
'Волнения', 'Волнения', 'Простуда'] * 4,
'Доза' : np.random.randint(low=1, high=6,size=24),
'Продолжительность' : np.random.randint(low=1, high=6,size=24)
})
df
```

```python
pd.crosstab(df['Врач'], df['Диагноз'], margins=True) # crosstab по умолчанию считает количество строк
```

```python
pd.crosstab(df['Врач'], df['Диагноз'])
```

```python
pd.crosstab(df['Врач'], df['Диагноз'], values=df['Доза'], aggfunc="mean")
```

```python
pd.pivot_table(df, index=['Врач'], values=['Доза', 'Продолжительность']) #pivot_table по умолчанию считает среднее
```

```python
pd.pivot_table(df,
    values='Доза',
    index=['Специальность', 'Врач'],
    columns=['Диагноз'],
    aggfunc="sum")
```

```python
pd.pivot_table(df,
    values=['Доза', 'Продолжительность'], # какие данные
    index=['Специальность', 'Врач'], #что в строках
    columns=['Диагноз'],# что в колонках
    aggfunc=["min", "mean", "max"], # что вычисляем
    margins=True).T #margins - это итоги
```

```python
pd.pivot_table(df,
    values=['Доза', 'Продолжительность'], # какие данные
    index=['Специальность', 'Врач'], #что в строках
    columns=['Диагноз'],# что в колонках
    aggfunc=["min", "mean", "max"], # что вычисляем
    margins=True).T #margins - это итоги

?pd.pivot_table
```
