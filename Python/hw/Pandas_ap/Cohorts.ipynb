```python
import pandas as pd

df = pd.read_csv('https://gist.githubusercontent.com/andron23/f8f772879116d9743f96917bd2c1ee24/raw/67a2c1e793e61c67bd55b5bc9220e5cc9602cd12/clients.csv')
df
```

```python
df.info()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
df.info()
```

```python
df[df['extra_leads'] == '{17089011}'].sort_values('created_at')
```

```python
df['first_dt'] = df.groupby('extra_leads')['created_at'].transform('min')
df['first_ym'] = df['first_dt'].dt.strftime('%Y-%m')
df['ym'] = df['created_at'].dt.strftime('%Y-%m')
df
```

```python
# df.groupby(['extra_leads'])['created_at'].transform('min')
```

```python
cohorts = pd.pivot_table(
    df,
    index='first_ym',
    columns='ym',
    values='extra_leads',
    aggfunc='nunique'
).fillna(0).astype(int)

cohorts
```

```python
import numpy as np
```

```python
diags = np.diag(cohorts)
cohorts = cohorts.div(diags, axis=0)
cohorts
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

f, ax = plt.subplots(figsize=(15, 6))

sns.heatmap(
    cohorts, 
    annot=True,
    fmt='.0%',
    linewidth=0.5,#медианное значение
    vmin=0.4,
    vmax=np.percentile(cohorts, 90),
    # cmap = 'spring'
);

ax.xaxis.set_label_position('top');
ax.xaxis.tick_top();

plt.xticks(rotation=45, ha='left');
```
