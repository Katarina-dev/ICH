```python
# 1. Создать класс Car (машина) со следующими полями: model, year, color.
# 2. Создать 10 объектов этого класса, описывающих модели разных марок, лет и
# цветов.
# 3. Создать список из этих объектов.
# 4. Написать функцию, которая принимает список объектов класса Car и цвет и
# возвращает список машин этого цвета. Напечатать этот список, выводя
# название модели, год и цвет. Использовать filter и lambda функции.


from datetime import datetime

class Car:
    def __init__(self, model, year, color):
        if not isinstance(year, int):
            raise TypeError('Недопустимый тип данных для года выпуска')
        if year < 1900 or year > datetime.now().year:
            raise ValueError('Недопустимый год выпуска')
        self.model = model
        self.year = year
        self.color = color
    def __str__(self): 
        return f"{self.model}, {self.year}, {self.color}"
    

cars = [
    Car("Toyota Corolla", 2020, "Red"),
    Car("Tesla", 2022, "Red"),
    Car("Honda", 2018, "Blue"),
    Car("KIA", 2013, "White"), 
    Car("Geely", 2012, "Black"),
    Car("Alfa Romeo", 2020, "Black"),
    Car("BMW", 2018, "Blue"),
    Car("Lada", 2013, "Red"), 
    Car("GM", 2012, "Black"),
    Car("Mercedes", 2020, "Blue"),
    Car("Ford Focus", "2018", "Blue"),
    Car("Invalid Car", "abcd", "White"), 
    Car("Future Car", 2030, "Black"), 
     ]

def filter_by_color(cars, color):
    return list(filter(lambda car: car.color.lower() == color.lower(), cars))
    
filtered_car = filter_by_color(cars, 'Red')
print('Машины красного цвета: ')

for car in filtered_car:
    print(car)



        
```

```python
# 5. Создать класс Person с полями имя и дата рождения.
# 6. Создать 10 объектов этого класса с разными именами.
# 7. Создать класс Employee который содержит поле имя и возраст.
# 8. Написать функцию, которая из списка объекта класса Person создает список
# из объектов класса Employee, вычисляя возраст каждого Person по дате
# рождения. Подумать, где должна быть реализована функция, вычисляющая
# возраст по дате рождения. Варианты: в конструкторе класса Employee, в
# качестве глобальной функции, в качестве метода класса (какого?).
# Получившийся список должен содержать сотрудников, старше 18 лет.
# Использовать map и filter. У классов Person и Employee должны быть
# определены конструкторы. Реализация трансформации список персонов в
# сотрудников должна быть в одну строчку.
# 9. Вывести получившихся сотрудников на экран.
# 10. Используя функцию forAll() убедиться, что все сотрудники действительно
# старше 18 лет.
from datetime import datetime

class Person:
    def __init__(self, name, birthdate):
        self.name = name
        self.birthdate =datetime.strptime(birthdate, "%Y-%m-%d")
        
    def __str__(self):
        return f'{self.name}, {self.birthdate}'

people = [ Person("Alice", "2020-05-14"), 
          Person("Bob", "1985-10-22"), 
          Person("Charlie", "1992-03-18"), 
          Person("Diana", "1995-07-29"), 
          Person("Eve", "1998-01-11"), 
          Person("Frank", "1987-12-05"), 
          Person("Grace", "1991-04-23"), 
          Person("Hank", "1989-08-15"), 
          Person("Ivy", "1996-02-09"), 
          Person("Jack", "1993-11-19") ]

class Employee:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def __str__(self):
        return f'{self.name}, {self.age}'

# for person in people:
#     print(person)


def person_age(person):
    today = datetime.today()
    age = today.year - person.birthdate.year - ((today.month, today.day) < (person.birthdate.month, person.birthdate.day))
    return age

def person_to_employee(person):
    return Employee(person.name, person_age(person))


result = list(map(person_to_employee, filter(lambda item: person_age(item) >= 18, people)))
for item in result:
    print(item)

def forAll(result_employee, condition):
    return all(condition(item) for item in result_employee)

if forAll(result, lambda e: e.age > 18):
    print("\nВсе сотрудники действительно старше 18 лет.")
else:
    print("\nЕсть сотрудники младше 18 лет.")
```
