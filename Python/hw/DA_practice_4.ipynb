```python
Напишите программу для генерации DataFrame с информацией о 35 сотрудниках компании. 
Фамилии, должности и отделы должны выбираться случайным образом из заранее заданного списка. 
Зарплата генерируется как случайное число с нормальным распределением.
Необходимо определить минимальную,максимальную и медианную зарплату. 
Вывести данные о сотрудниках с найденными зарплатами.
```

```python
surnames = ['Walger', 'Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia', 'Miller', 'Davis', 'Martinez', 'Taylor']
positions = ['Manager', 'Developer', 'Analyst', 'Administrator', 'Designer', 'Tester', 'Consultant']
departments = ['Sales', 'IT', 'Marketing', 'Accounting', 'HR', 'Logistics']
n = 35
np.random.seed(42)

dic = {'surnames': np.random.choice(surnames, n),
       'positions': np.random.choice(positions, n),
       'departments': np.random.choice(departments, n),
       'salary': np.random.normal(10000, 8000, size = n)
      }
dic

df = pd.DataFrame(dic)


df.loc[df['salary'] < 1000, 'salary'] = 1000 #формирует срез, выбирает все salary меньше 1000, берем колонку salary и присваиваим 1000 таким значениям, так работаем с выбросами
display(df)

print(f'Минимальная зарплата: {df['salary'].min()}')
print(f'Максимальная зарплата: {df['salary'].max()}')
print(f'Медианная зарплата: {df['salary'].median()}')

salary_max = df['salary'].max()
#salary_max = df.salary.max() так тоже можно писать, но в этом случае колонки не должны содержать пробелы и русские буквы. Нужно когда часто используем колонки - быстрый доступ
df[df.salary == salary_max]
```

```python
# вариант Насти
surnames = ['Walger','Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia', 'Miller', 'Davis', 'Martinez', 'Taylor']
positions = ['Manager', 'Developer', 'Analyst', 'Administrator', 'Designer', 'Tester', 'Consultant']
departments = ['Sales', 'IT', 'Marketing', 'Accounting', 'HR', 'Logistics']
n = 35

np.random.seed(42)

dic = {
    'surname' : np.random.choice(surnames, n),
    'position' : np.random.choice(positions, n),
    'department' : np.random.choice(departments, n),
    'salary' : np.random.normal(10000, 8000, size = n)
}

df = pd.DataFrame(dic)
#display(df)


df.loc[df['salary'] < 1000, 'salary'] = 1000
display(df)

print(f'минимальная зарплата: {df['salary'].min()}')
print(f'максимальная зарплана: {df['salary'].max():.2f}')
print(f'медиана: {df['salary'].median():.2f}')
```
