```python
1. Работа с временными рядами в pandas:


Напишите программу, которая:

- Создаёт DataFrame с двумя столбцами: "Дата" (в формате дд-мм-гггг) и "Значение" (целые числа).

- Преобразует столбец Дата в тип datetime и устанавливает его в качестве индекса DataFrame.

- Генерирует временной ряд, заполняя столбец Значение случайными числами.

- Выводит сгенерированный временной ряд.

- Запрашивает у пользователя начальную и конечную дату периода для фильтрации.

- Выводит временной ряд за указанный период.
```

```python
import numpy as np
import pandas as pd

cols = ['Data', 'Value']
my_df = pd.DataFrame(columns=cols)
my_df['Data'] = pd.to_datetime(my_df['Data'])
my_df.set_index('Data', inplace = True)

dates = pd.date_range(start='01-01-2023', end='31-12-2023', freq='D')
values = np.random.randint(0, 500, size=len(dates))
my_df = my_df.reindex(dates)

my_df['Value'] = values

my_df.index = my_df.index.strftime('%d-%m-%Y')  

my_df.index.name = 'Data'
# print(my_df)
my_df.index
display(my_df)
start_date = input('Enter start date (dd-mm-YYYY): ')
end_date = input('Enter end date in format (dd-mm-YYYY): ')
print(my_df.loc[start_date:end_date ])

```

```python
2. Работа с несколькими DataFrame


Напишите программу, которая:

- Создаёт DataFrame с тремя столбцами: "Товар", “Откуда”, “Куда” (в последних двух столбцах указываются города).

- Создаёт второй DataFrame с двумя столбцами: "Город", “Страна”.

- На основе слияния двух предыдущих DataFrame создает третий, в котором для каждого товара вместо городов доставки 
будут указаны страны. Лишних столбцов в этом DataFrame быть не должно, только "Товар", “Откуда”, “Куда” 
(в последних двух столбцах указываются страны)

- Добавьте в последний DataFrame новый столбец “Внутри_страны”, который принимает логические значения (True, если доставка осуществляется внутри страны).

- Вычислите, сколько в наших данных доставок внутри страны. 
```

```python
goods = ['Apples', 'Bananas', 'Carrots', 'Milk']
cities = ['New York', 'London', 'Tokyo', 'Novi Sad', 'Paris', 'Berlin', 'Amsterdam', 'Belgrade']
countries = ['United States', 'United Kingdom', 'Japan', 'Serbia',  'France', 'Germany', 'Netherlands', 'Serbia']

df_1 = pd.DataFrame({'Goods': goods, 'From': cities[:4], 'To': cities[4:]})
df_2 = pd. DataFrame({'City': cities, 'Country': countries})

display(df_1)
display(df_2)

df_3 = df_1.merge(df_2, left_on = 'From', right_on = 'City', how = 'inner').rename(columns={'Country': 'Country_from'})
df_4 = df_1.merge(df_2, left_on = 'To', right_on = 'City', how = 'inner').rename(columns={'Country': 'Country_to'})
display(df_3)
display(df_4)

df_5 = df_3[['Goods', 'Country_from']].merge(df_4[['Goods', 'Country_to']], on='Goods')
df_5['Domestic delivery'] = df_3['Country_from'] == df_4['Country_to']
display(df_5)
print(f'Domestic delivery count: {df_5['Domestic delivery'].sum()}')                 
```

```python
Напишите программу, которая:

Запрашивает у пользователя количество строк данных для ввода.

Для каждой строки запрашивает данные по трем столбцам: "Имя", "Возраст", "Регион", разделённые пробелом.

Создает DataFrame на основе введенных данных.

Выводит общую информацию о DataFrame, включая количество записей, средний возраст, и количество уникальных регионов.

Выбирает и выводит на экран данные о самом молодом участнике.

```

```python
count_str = int(input('Enter count of string: '))

data = []
for i in range(count_str):
    query = input('Enter Name, Age, Region (with spaces): ').split()
    data.append(query)
df = pd.DataFrame(data, columns = ['Name', 'Age', 'Region'])
df['Age'] = pd.to_numeric(df['Age'])
youngest = (df[df['Age'] == df['Age'].min()].iloc[0]) 

display(df)
print(f"Count of rows: {df.shape[0]}\n"
    f"Average age: {df['Age'].mean():.2f}\n"
    f"Count of unique region: {df['Region'].nunique()}\n"
    f"The yongest participant is:\n {' '.join(youngest.astype(str))}")
    
```
