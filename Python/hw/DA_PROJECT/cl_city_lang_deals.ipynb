```python
# Регулярное выражение для поиска уровней
level_pattern = r'([A-C][1-2]|[А-Я][1-2])(?:-([A-C][1-2]|[А-Я][1-2]))?'

# Словарь для перевода уровней на английский
level_translation = {
    'A0': 'A0',
    'A1': 'A1',
    'A2': 'A2',
    'B1': 'B1',
    'B2': 'B2',
    'C1': 'C1',
    'C2': 'C2',
    'А1': 'A1',
    'А1': 'A1',
    'А2': 'A2',
    'Б1': 'B1',
    'Б2': 'B2',
    'Ц1': 'C1',
    'Ц2': 'C2',
}

# Регулярное выражение для поиска уровней
# Учитываем как латиницу, так и кириллицу, а также комбинации уровней типа 'A2-B1'
level_pattern = r'([A-C][0-2]|[А-Я][0-2])(?:-([A-C][0-2]|[А-Я][0-2]))?'

#нормализация кириллических букв
def normalize_cyrillic(text):
    return (str(text)
        .replace('А', 'A')
        .replace('В', 'B')
        .replace('С', 'C')
        .replace('Б', 'B')
        .replace('Ц', 'C'))

    
# Функция для извлечения уровня и перевода
def extract_level(text):
    text = normalize_cyrillic(str(text).upper())  # сначала нормализуем
    match = re.search(level_pattern, text)
    if match:
        if match.group(2):
            return level_translation.get(match.group(1), match.group(1)) + '-' + level_translation.get(match.group(2), match.group(2))
        else:
            return level_translation.get(match.group(1), match.group(1))
    return None
```

```python
# Применяем функцию к колонке 'Level of Deutsch'
deals_df_v3['Level of Deutsch Cleaned'] = deals_df_v3['Level of Deutsch Cleaned'].apply(normalize_cyrillic).apply(extract_level)
```

```python
import numpy as np

# Categorize SLA
def sla_category(td):
    if pd.isna(td): # 1. Проверка: если значение пустое (NaT или NaN)
        return np.nan #    → возвращаем np.nan
    minutes = td.total_seconds() / 60 # 2. Переводим timedelta в минуты
    # 3. Определяем, в какой диапазон попало время
    if minutes <= 5:
        return 'Very fast (≤5 min)'
    elif minutes <= 15:
        return 'Fast (5-15 min)'
    elif minutes <= 30:
        return 'Medium (15-30 min)'
    elif minutes <= 60:
        return 'Slow (30-60 min)'
    else:
        return 'Very slow (>60 min)'

deals_df_v3['SLA_range'] = deals_df_v3['SLA'].apply(sla_category)
```
