```python
import numpy as np

# Сначала сортируем по времени
sorted_df = duplicates_contact_name.sort_values('Created Time')

# Указываем, что делать для каждого столбца
aggregated_contacts = (
    sorted_df
    .groupby('Contact Name')
    .agg(lambda x: (
        x.iloc[0] if x.name == 'Source'  # для Source — первое значение (раньше по времени)
        else (
            x.dropna()[x.dropna() != 0].iloc[-1]  # последнее ненулевое
            if not x.dropna()[x.dropna() != 0].empty
            else x.dropna().iloc[-1]              # или последнее непустое
            if not x.dropna().empty
            else np.nan                           # иначе NaN
        )
    ))
    .reset_index()
)
```

```python
# 6. Удаляем старые строки с этими Contact Name
mask_to_remove = (
    (deals_df_v2['Stage'] == 'Lost') &
    (deals_df_v2['Lost Reason'] == 'Duplicate') &
    (deals_df_v2['Contact Name'].isin(duplicate_ids_more_one))
)
deals_df_v2 = deals_df_v2[~mask_to_remove].copy()
deals_df_v2.shape[0]
```
