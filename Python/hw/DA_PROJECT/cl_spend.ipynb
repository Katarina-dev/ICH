```python
# Шаг 1: Оставляем строки, где хотя бы одна метрика не равна нулю, т.к. если пусты все эти параметры, то строка не несет для нас пользы, остальные выделяем
# Сохраняем строки для удаления в отдельный DataFrame
removed_rows = spend_df[
    (spend_df['Spend'] == 0) & 
    (spend_df['Clicks'] == 0) & 
    (spend_df['Impressions'] == 0)
]
```

```python
# Шаг 3: Удаляем строки с нулевыми значениями из spend_df
# Используем .loc, чтобы изменить исходный DataFrame, избегая предупреждений
spend_df = spend_df.loc[
    (spend_df['Spend'] != 0) | 
    (spend_df['Clicks'] != 0) | 
    (spend_df['Impressions'] != 0)
].copy()
```

```python
#  Очистим и приведем к единому формату следующие текстовые столбцы
spend_df['Source'] = spend_df['Source'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
spend_df['Campaign'] = spend_df['Campaign'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
spend_df['AdGroup'] = spend_df['AdGroup'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
spend_df['Ad'] = spend_df['Ad'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
spend_df
```

```python
# Теперь работаем с spend_df и изменяем типы данных
spend_df['Source'] = spend_df['Source'].astype('category')
spend_df['Campaign'] = spend_df['Campaign'].astype('category')
spend_df['AdGroup'] = spend_df['AdGroup'].astype('category')
```
