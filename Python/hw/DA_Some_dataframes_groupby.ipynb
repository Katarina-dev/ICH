```python
1. У вас есть датафрейм с данными о продажах различных товаров в разных
регионах. Необходимо сгруппировать данные по регионам и вычислить общее
количество проданных товаров в каждом регионе.
```

```python
sales_df = pd.DataFrame({'region': ['Region A', 'Region B', 'Region A', 'RegionB'],
'product': ['Product X', 'Product Y', 'Product X','Product Y'],
'quantity': [100, 150, 200, 250]})
sales_df
```

```python
sales_df.groupby('region')['quantity'].sum()
```
