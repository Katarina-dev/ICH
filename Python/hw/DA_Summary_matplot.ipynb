```python
Напишите программу, которая генерирует случайные данные об объеме выручки в некотором магазине по месяцам. 
Нарисуйте график по этим данным.
Постройте круговую диаграмму.
Постройте столбчатую диаграмму.
```

```python
np.random.seed(42)
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
revenue = np.random.randint(5000, 20000, size=12)

df = pd.DataFrame({'Month': months, 'Revenue': revenue})

plt.plot(months, revenue, marker='o') # первое x, второе y
plt.title("Revenue by months")
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid()

for i, txt in enumerate(revenue):
    plt.text(months[i], revenue[i] + 500, str(txt), ha='center', fontsize=10, color='black')

plt.show()
```

```python
df = pd.DataFrame({'Month': months, 'Revenue': revenue}) 
plt.plot(df.Month, df.Revenue) # из датафрейма
plt.title("Revenue by months")
plt.xticks(rotation=45)
for i, value in enumerate(df['Revenue']):
    plt.annotate(value, (df['Month'][i], df['Revenue'][i]), textcoords="offset points", xytext=(0, 10), ha='center')
plt.show()
```

```python
plt.pie(df.Revenue, labels=df.Month, autopct='%1.1f%%', startangle=90)
plt.title("Revenue by months")
plt.show()
```

```python
plt.bar(df.Month, df.Revenue, color='violet', edgecolor='black')
plt.title("Revenue by months")
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
В предыдущей задаче сгенерируйте два аналогичных набора данных – информацию о выручке еще двух магазинов. 
На одной системе координат изобразите все три графика.
Постройте столбчатую диаграмму для сравнения выручки магазинов.
```

```python
np.random.seed(42)
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
revenue = np.random.randint(5000, 20000, size=12)
revenue_1 = np.random.randint(5000, 20000, size=12)
revenue_2 = np.random.randint(5000, 20000, size=12)

df = pd.DataFrame({'Month': months, 'Revenue': revenue})
df_1 = pd.DataFrame({'Month': months, 'Revenue': revenue_1})
df_2 = pd.DataFrame({'Month': months, 'Revenue': revenue_2})

plt.plot(df.Month, df.Revenue, marker='o', label= 'shop_0')
plt.plot(df_1.Month, df_1.Revenue, marker='d', label= 'shop_1')
plt.plot(df_2.Month, df_2.Revenue, marker='*', label= 'shop_2') # первое x, второе y
plt.title("Revenue by months")
plt.xticks(rotation=45)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid()

for i, txt in enumerate(revenue):
    plt.text(months[i], revenue[i] + 500, str(txt), ha='center', fontsize=10, color='black')
plt.legend()
plt.show()
```

```python
w = 0.25
x = np.arange(12)

plt.bar(x-w, df.Revenue, color='violet', edgecolor='black', width=w)
plt.bar(x, df_1.Revenue, color='blue', edgecolor='black', width=w)
plt.bar(x+w, df_2.Revenue, color='pink', edgecolor='black', width=w)
plt.title("Revenue by months")
plt.xticks(x, months, rotation=45)
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```
