```python
1.Совмещение двух графиков

Изобразите на одной системе координат графики функций y=sin(x) и y=log(x+1). 
Используйте две оси Y: одну для первого графика, и вторую для другого.
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.arange(-100, 100)
x2 = np.arange(1, 100)
y1 = np.sin(x1)
y2 = np.log(x2+1)

plt.figure()
ax1 = plt.subplot()
ax1.plot(x1, y1, color='green')
ax1.tick_params(axis='y', labelcolor='green')

ax2 = ax1.twinx()
ax2.plot(x2, y2, color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.show()
```

```python
2.В датафрейме ниже представлены результаты соревнований: для каждого игрока указано  набранное количество очков в 
нескольких попытках.

data = {

    'Name': ['Анна', 'Антон', 'Мария', 'Макс', 'Макс', 'Алекс', 'Антон', 'Мария', 'Макс', 'Алекс', 'Антон', 'Анна'],

     'Points': [145, 2524, 343, 4212, 5212, 6421, 745, 842, 524, 3224, 212, 1214]      

}

Создайте из этих данных датафрейм. Сгруппируйте данные по имени игроков. 
Методом plot() для датафреймов постройте график минимальных, максимальных и средних очков каждого игрока на 
одной системе координат.
```

```python
data = pd.DataFrame({'Name': ['Анна', 'Антон', 'Мария', 'Макс', 'Макс', 'Алекс', 'Антон', 'Мария', 'Макс', 'Алекс', 'Антон', 'Анна'],
     'Points': [145, 2524, 343, 4212, 5212, 6421, 745, 842, 524, 3224, 212, 1214]})

plt.figure(figsize=(15, 9))

min_points = data.groupby('Name')['Points'].min()
max_points = data.groupby('Name')['Points'].max()
mean_points = data.groupby('Name')['Points'].mean()

plt.plot (min_points, color='green', marker='o', label='Minimum')
plt.plot(max_points, color='orange', marker='o', label='Maximum')
plt.plot(mean_points, color='red', marker='o', label='Mean')
plt.xlabel('Name')
plt.ylabel('Points')
plt.legend()
plt.grid(True)

for i, name in enumerate(min_points.index):
    plt.text(name, min_points[name], str(min_points[name]), ha='center')
    plt.text(name, max_points[name], str(max_points[name]), ha='center')
    plt.text(name, mean_points[name], str(round(mean_points[name], 2)), ha='center')

plt.show()
```
