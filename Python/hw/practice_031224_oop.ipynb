```python
Задание 1: Создайте классы Product и Cart для моделирования процесса покупок в интернет-магазине.

Класс Product
Атрибуты:
name: название продукта.
price: цена продукта.
category: категория продукта (например, "electronics", "furniture").
метод __str__, который возвращает строковое представление объекта Product.
метод __repr__, который возвращает строку, удобную для отладки / десериализации

Класс Cart
Атрибуты:
id: id корзины
items: словарь объектов типа Product: quantity, которые добавлены в корзину.
status: статус корзины (например, "active", "completed", "canceled").
метод add(product, quantity): добавляет продукт в корзину. Если продукт уже есть в корзине, увеличивает количество.
метод remove(product, quantity): удаляет продукт из корзины.
total_price(): возвращает общую сумму всех товаров в корзине с учетом их количества.
__str__: возвращает строковое представление корзины, показывающее список всех товаров с количеством и суммой.
```

```python
 class Product:
    def __init__(self, name, price, category):
        self.name = name
        self.price = price
        self.category = category

    def __str__(self):
        return f'{self.name},{self.price},{self.category}'
    def __repr__(self): 
        return f'{self.__dict__}'
        
class Cart:
    def __init__(self, id):

        self.id = id
        self.items = {}
        self.status = "active"

def add(self, product, quantity):
        if product in self.items:
            self.items[product] += quantity
        else:
            self.items[product] = quantity
     
def remove(self, product, quantity):
        if product in self.items:
            self.items[product] -= quantity
            if self.items[product] <= 0:
                del self.items[product]
def total_price(self):
        return sum(product.price * quantity for product, quantity in self.items.items())
def __str__(self):
        item_list = [f"{product.name}: {product.price} x {quantity} = {product.price * quantity} руб." for product, quantity in self.items.items()]
        items_str = "\n".join(item_list)
        return f"Cart ID: {self.id}\nStatus: {self.status}\nItems:\n{items_str}\nTotal: {self.total_price()} руб."

pr1 = Product("Laptop", 1500, "electronics")
pr2 = Product("Chair", 300, "furniture")



cart = Cart(1) 
cart.add(pr1, 2) 
cart.add(pr2, 1)

print(cart)
```

```python
def calculate_total_revenue(carts):
    total_revenue = 0
    for cart in carts:
        total_revenue += cart.total_price()
    return total_revenue

def calculate_average_cart_value(carts):
    total_revenue = calculate_total_revenue(carts)
    average_cart_value = total_revenue / len(carts) if carts else 0
    return average_cart_value

def filter_carts_by_status(carts, status):
    filtered_carts = [cart for cart in carts if cart.status == status]
    return filtered_carts

def filter_carts_by_category(carts, category):
    filtered_carts = []
    for cart in carts:
        for product in cart.items.keys():
            if product.category == category:
                filtered_carts.append(cart)
                break
    return filtered_carts

product1 = Product("Laptop", 1500, "electronics")
product2 = Product("Chair", 300, "furniture")
product3 = Product("Tablet", 500, "electronics")

cart1 = Cart(1)
cart1.add(product1, 2)  
cart1.add(product2, 1)  

cart2 = Cart(2)
cart2.add(product3, 3)  
cart2.add(product2, 2)  
cart2.complete()

cart3 = Cart(3)
cart3.add(product1, 1)  
cart3.add(product3, 1) 
cart3.cancel()

carts = [cart1, cart2, cart3]

total_revenue = calculate_total_revenue(carts)
print(f"Общий доход: {total_revenue}")

average_cart_value = calculate_average_cart_value(carts)
print(f"Средняя стоимость корзины: {average_cart_value}")

completed_carts = filter_carts_by_status(carts, "completed")
print(f"Завершенные корзины: {completed_carts}")


electronics_carts = filter_carts_by_category(carts, "electronics")
print(f"Корзины с товарами категории 'electronics': {electronics_carts}")
```
