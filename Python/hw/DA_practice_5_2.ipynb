```python
В некотором учебном заведении стобальная система оценивания. Баллы от 0 до 20 – это “неудовлетворительно”, от 21 до 55 
– “посредственно”, от 56 до 70 – “удовлетворительно”, от 71 до 90 – “хорошо” и от 91 до 100 – “отлично”. 
Создайте датафрейм из трех столбцов: словесная оценка, минимальный балл, максимальный балл.
Создайте второй датафрейм из одного столбца, заполненного случайными оценками от 0 до 100. Соедините созданные датафреймы.
```

```python
import pandas as pd
import numpy as np

# 1 вариант
estimates = pd.DataFrame({'word_est': ['неудовлетворительно', 'посредственно', 'удовлетворительно', 'хорошо', 'отлично'],
                          'min_est': [0, 21, 56, 71, 91],
                          'max_est': [20, 55, 70, 90, 100]})
display(estimates)
np.random.seed(42)
random_scores = pd.DataFrame({'Score': np.random.randint(0, 101, 20)})

def get_grade(score, estimates):
    for _, row in estimates.iterrows():
        if row['min_est'] <= score <= row['max_est']:
            return row['word_est']

random_scores['word_est'] = random_scores['Score'].apply(lambda x: get_grade(x, estimates))
        


display(random_scores)
```
