```python
import numpy as np

cat_cols=deals_df.select_dtypes(include=['object', 'category']).columns.tolist()
num_cols = deals_df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


for col in num_cols:
    print(col)
    plt.figure(figsize=(15, 4))
    
    # Если столбец timedelta — преобразуем его в дни
    series = deals_df[col]
    if pd.api.types.is_timedelta64_dtype(series):
        series = series.dt.total_seconds() / 3600  # переводим в часы

    plt.subplot(1, 2, 1)
    series.hist(grid=False)
    plt.ylabel('count')

    plt.subplot(1, 2, 2)
    sns.boxplot(x=series)
    plt.show()
```

```python
light_cat_cols = [
    col for col in deals_df.select_dtypes(include=['category']).columns
    if deals_df[col].nunique() <= 20
]

subset = light_cat_cols  # максимум 8 графиков
n_cols = 2
n_rows = (len(subset) + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 6 * n_rows))
axes = axes.reshape(n_rows, n_cols)  # на случай одномерного массива

for i, col in enumerate(subset):
    row = i // n_cols
    col_idx = i % n_cols
    ax = axes[row, col_idx]

    # Данные без NaN
    data = deals_df[col].dropna()
    
    sns.countplot(x=data, color='blue', ax=ax)

    total = len(data)
    for p in ax.patches:
        height = p.get_height()
        percent = f'{100 * height / total:.1f}%'
        ax.annotate(percent, 
                    (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom', fontsize=10, color='black')

    ax.set_title(col)
    ax.tick_params(axis='x', labelrotation=45)

# Удалим пустой график, если их нечётное количество
if len(subset) % n_cols != 0:
    fig.delaxes(axes[-1, -1])

plt.tight_layout()
plt.show()
```

```python
# Убираем NaN и пустые строки
filtered_cam_df = deals_df[deals_df['Campaign'].notna() & (deals_df['Campaign'] != '')]

# Получаем порядок топ-10 кампаний по убыванию
top_campaigns = filtered_cam_df['Campaign'].value_counts().head(10).index

# Оставляем только топ-10 для отображения
top_df = filtered_cam_df[filtered_cam_df['Campaign'].isin(top_campaigns)]

# Строим график
plt.figure(figsize=(10, 6))
sns.countplot(data=top_df, x='Campaign', order=top_campaigns, color='blue')
plt.xticks(rotation=45, ha='right')  # ha='right' подравнивает подписи
plt.title('Top-10 Campaigns by count')
plt.xlabel('Campaign')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(17, 17))

for k, col in enumerate(spend_cat_cols):
    ax = axes[k // 2, k % 2]

    # Удалим NaN
    data = spend_df[spend_df[col].notna()].copy()

    # Для Campaign, AdGroup, Ad: объединяем редкие в "Other"
    if col in ['Campaign', 'AdGroup', 'Ad']:
        top_10 = data[col].value_counts().nlargest(10).index
        data[col] = data[col].apply(lambda x: x if x in top_10 else np.nan)

    # Сортировка по убыванию
    order = data[col].value_counts().index

    sns.countplot(ax=ax, x=col, data=data, color='blue', order=order)
    ax.set_title(col)
    ax.tick_params(axis='x', labelrotation=70)

    # Подписи в процентах
    total = len(data)
    for p in ax.patches:
        height = p.get_height()
        percent = 100 * height / total
        ax.annotate(f'{percent:.1f}%', (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom', fontsize=10)

fig.suptitle('Top-10 by count')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

for ca_col in calls_num_cols:
    print(ca_col)

    # Фильтруем звонки с длительностью больше 3 секунд, если это нужный столбец
    if ca_col == 'Call Duration (in seconds)':
        data = calls_df[ca_col]
        data = data[data > 3].dropna()
    else:
        data = calls_df[ca_col].dropna()

    plt.figure(figsize=(15, 4))

    if ca_col == 'Call Duration (in seconds)':
        plt.hist(data[data <= 2000], bins=30, color='blue', alpha=0.7, label='<= 1000 sec')
        plt.hist(data[data > 2000], bins=5, color='red', alpha=0.7, label='> 1000 sec')
        plt.legend()
        plt.ylabel('count')
        plt.title(f'Histogram of {ca_col} with "> 1000 sec" highlighted')
    else:
        data.hist(grid=False)
        plt.ylabel('count')
        plt.title(f'Histogram of {ca_col}')

    plt.show()
```
