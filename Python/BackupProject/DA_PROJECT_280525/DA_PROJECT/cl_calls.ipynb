```python
def categorize_duration(sec):
    if pd.isna(sec):
        return np.nan
    elif sec == 0:
        return '0 sec'
    elif 1 <= sec <= 3:
        return '1–3 sec'
    elif sec > 3:
        # Переводим секунды в минуты
        minutes = sec / 60
        if minutes < 0.25:      # < 15 сек
            return '3–15 sec'
        elif minutes < 1:       # 15–60 сек
            return '15–60 sec'
        elif minutes < 3:       # 1–3 мин
            return '1–3 min'
        elif minutes < 10:      # 3–10 мин
            return '3–10 min'
        elif minutes < 30:      # 10–30 мин
            return '10–30 min'
        else:                  # >30 мин
            return '30+ min'
    else:
        return 'Other'

calls_df['Call Duration Range'] = calls_df['Call Duration (in seconds)'].apply(categorize_duration)
```
