```python
# Output for analyzing Categorical Variables and Numerical Variables

import numpy as np

cat_cols=deals_df.select_dtypes(include=['object', 'category']).columns.tolist()
num_cols = deals_df.select_dtypes(include=np.number).columns.tolist()
print("Categorical Variables:")
print(cat_cols)
print("Numerical Variables:")
print(num_cols)
```

```python
# Mode
mode_values = deals_df[num_cols].mode().iloc[0]
mode_values
```

```python
mode_cat_values = deals_df[cat_cols].mode().iloc[0]
mode_cat_values
```

```python
# Range
range_values = deals_df[num_cols].max() - deals_df[num_cols].min()
range_values
```

```python
# Make some introductory graphs for numerical variables

import matplotlib.pyplot as plt
import seaborn as sns


for col in num_cols:
    print(col)
    plt.figure(figsize=(15, 4))
    
    # If the column is timedelta, convert it to days
    series = deals_df[col]
    if pd.api.types.is_timedelta64_dtype(series):
        series = series.dt.total_seconds() / 86400  # convert to days

    plt.subplot(1, 2, 1)
    series.hist(grid=False)
    plt.ylabel('count')

    plt.subplot(1, 2, 2)
    sns.boxplot(x=series)
    plt.show()
```

```python
# Make some introductory graphs for categorical variables

light_cat_cols = [
    col for col in deals_df.select_dtypes(include=['category']).columns
    if deals_df[col].nunique() <= 20
]

subset = light_cat_cols 
n_cols = 2
n_rows = (len(subset) + 1) // n_cols

fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, 6 * n_rows))
axes = axes.reshape(n_rows, n_cols)  # for the case of a one-dimensional array

for i, col in enumerate(subset):
    row = i // n_cols
    col_idx = i % n_cols
    ax = axes[row, col_idx]

    # Данные без NaN
    data = deals_df[col].dropna()
    
    sns.countplot(x=data, color='blue', ax=ax)

    total = len(data)
    for p in ax.patches:
        height = p.get_height()
        percent = f'{100 * height / total:.1f}%'
        ax.annotate(percent, 
                    (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom', fontsize=10, color='black')

    ax.set_title(col)
    ax.tick_params(axis='x', labelrotation=45)

# Remove empty graph if there is an odd number of them
if len(subset) % n_cols != 0:
    fig.delaxes(axes[-1, -1])

plt.tight_layout()
plt.show()
```

```python
# Filter NaN and empty strings
filtered_cam_df = deals_df[deals_df['Campaign'].notna() & (deals_df['Campaign'] != '')]

# get the order of top 10 campaigns in descending order
top_campaigns = filtered_cam_df['Campaign'].value_counts().head(10).index

# leave only the top 10 for display
top_df = filtered_cam_df[filtered_cam_df['Campaign'].isin(top_campaigns)]

# Building a diagram
plt.figure(figsize=(10, 6))
sns.countplot(data=top_df, x='Campaign', order=top_campaigns, color='blue')
plt.xticks(rotation=45, ha='right')  # ha='right' aligns signatures
plt.title('Top-10 Campaigns by count')
plt.xlabel('Campaign')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, axes = plt.subplots(2, 2, figsize=(17, 17))

for k, col in enumerate(spend_cat_cols):
    ax = axes[k // 2, k % 2]

    # Remove NaN
    data = spend_df[spend_df[col].notna()].copy()

    # For Campaign, AdGroup, Ad: Combine rare ones into "Other"
    if col in ['Campaign', 'AdGroup', 'Ad']:
        top_10 = data[col].value_counts().nlargest(10).index
        data[col] = data[col].apply(lambda x: x if x in top_10 else np.nan)

    # Sort in descending order
    order = data[col].value_counts().index

    sns.countplot(ax=ax, x=col, data=data, color='blue', order=order)
    ax.set_title(col)
    ax.tick_params(axis='x', labelrotation=70)

    # Signatures in percentage
    total = len(data)
    for p in ax.patches:
        height = p.get_height()
        percent = 100 * height / total
        ax.annotate(f'{percent:.1f}%', (p.get_x() + p.get_width() / 2, height),
                    ha='center', va='bottom', fontsize=10)

fig.suptitle('Top-10 by count')
plt.tight_layout()
plt.show()
```

```python
# Analyze calls with duration more than 3 seconds

import numpy as np

for ca_col in calls_num_cols:
    print(ca_col)

    # Filter calls with duration more than 3 seconds, if this is the desired column
    if ca_col == 'Call Duration (in seconds)':
        data = calls_df[ca_col]
        data = data[data > 3].dropna()
    else:
        data = calls_df[ca_col].dropna()

    plt.figure(figsize=(15, 4))

    if ca_col == 'Call Duration (in seconds)':
        plt.hist(data[data <= 2000], bins=30, color='blue', alpha=0.7, label='<= 1000 sec')
        plt.hist(data[data > 2000], bins=5, color='red', alpha=0.7, label='> 1000 sec')
        plt.legend()
        plt.ylabel('count')
        plt.title(f'Histogram of {ca_col} with "> 1000 sec" highlighted')
    else:
        data.hist(grid=False)
        plt.ylabel('count')
        plt.title(f'Histogram of {ca_col}')

    plt.show()
```

```python
# Analyze calls by Call Duration Range

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
order = ['0 sec', '1–3 sec', '3–15 sec', '15–60 sec', '1–3 min', '3–10 min', '10–30 min', '30+ min']
sns.countplot(x='Call Duration Range', data=calls_df, order=order)
plt.title('Distribution of calls by duration')
plt.xlabel('Call duration')
plt.ylabel('Number of calls')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
