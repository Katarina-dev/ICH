```python
import pandas as pd

deals_df = pd.read_pickle("deals_cleaned_v3.pkl")
deals_df
```

```python
city_rename_dict = {
    'Neuburg': 'Neuburg an der Donau',
    'Bat yam': 'Bat Yam',
    'Halle': 'Halle (Saale)'
}
deals_df['City'] = deals_df['City'].replace(city_rename_dict)
```

```python
# Загрузка файла с координатами
city_coords_df = pd.read_csv("City Coordinates.csv")
city_coords_df
```

```python
# Убираем дубликаты городов — оставляем первую встречу
city_coords_df_unique = city_coords_df.drop_duplicates(subset='City')
city_coords_df_unique
```

```python

# Объединяем
deals_df = deals_df.merge(city_coords_df_unique[['City', 'Latitude', 'Longitude']], on='City', how='left')
deals_df
```

```python
not_matched = deals_df[
    deals_df['City'].notna() &  # Город указан
    (deals_df['Latitude'].isna() | deals_df['Longitude'].isna())  # Но координаты не подтянулись
]

print(f"Городов, не нашедших координаты: {len(not_matched)}")
print("Примеры:")
print(not_matched['City'].drop_duplicates().to_list()[:10])
```
